```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
train_X=np.load("../data/train_X.npy")
train_y=np.load("../data/train_y.npy")>0
test_X=np.load("../data/test_X.npy")
test_y=np.load("../data/test_y.npy")>0
```

### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.

# TO DO

# Solution:
positive_delta=2*sum(train_y)-len(train_y)
if positive_delta>0:
    extras=np.random.choice(np.where(train_y==0)[0],size=positive_delta,replace=True)
else:
    extras=np.random.choice(np.where(train_y>0)[0],size=-positive_delta,replace=True)
train_y=np.concatenate([train_y,train_y[extras]])
train_X=np.concatenate([train_X,train_X[extras]])
print(f'total: {len(train_y)}, positive: {sum(train_y)}')
permutation=np.random.permutation(range(len(train_y)))
val_folds=[permutation[i::5] for i in range(5)]
train_folds=[[x for x in permutation if x not in val_fold] for val_fold in val_folds]
folds=[((train_X[train_folds[i]],train_y[train_folds[i]]),(train_X[val_folds[i]],train_y[val_folds[i]])) for i in range(5)]
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
def compute_loss(X,y,theta,lam):
    
    # TO DO
    
    # Solution:
    return -sum([y_i*np.log(sigmoid(np.dot(theta,X_i)))+(1-y_i)*np.log(1-sigmoid(np.dot(theta,X_i))) for X_i,y_i in zip(X,y)])+lam*np.matmul(theta,np.transpose(theta))
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        
        # TO DO
        
        # Solution:
        term_1=sigmoid(np.matmul(self.theta,np.transpose(X)))
        term_2=np.matmul(y,X)
        return np.matmul(term_1,X)-term_2+2*self.lam*self.theta
    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.

def validate(models,XXs,yys,label,axis):
    
    # TO DO
    
    # Solution:
    auc=[]
    sensitivity,specificity=[],[]
    for i,model in enumerate(models):
        sensitivity.append([1])
        specificity.append([0])
        auc.append(0)
        predictions=model.proba(XXs[i])
        yys[i]=yys[i][np.argsort(predictions)]
        predictions=sorted(predictions)
        for j,prediction in enumerate(predictions):
            if yys[i][j]>0:
                specificity[i].append(specificity[i][-1])
                sensitivity[i].append(sensitivity[i][-1]-1/sum([y>0 for y in yys[i]]))
            else:
                sensitivity[i].append(sensitivity[i][-1])
                specificity[i].append(specificity[i][-1]+1/sum([y==0 for y in yys[i]]))
                auc[i]+=(specificity[i][-1]-specificity[i][-2])*sensitivity[i][-1]
        axis.plot([1-specif for specif in specificity[i]],sensitivity[i],color='pink',linewidth=1.5,alpha=0.7)
    currs=[1]*len(models)
    alive=list(range(len(models)))
    specificity_for_avg,sensitivity_for_avg=[],[]
    while len(alive)>0:
        candidates=[specificity[i][currs[i]] for i in alive]
        minimal_i=alive[np.argmin(candidates)]
        specificity_for_avg.append(specificity[minimal_i][currs[minimal_i]])
        sensitivity_for_avg.append(np.mean([sensitivity[i][currs[i]-1] for i in range(len(models))]))
        currs[minimal_i]+=1
        if currs[minimal_i]==len(sensitivity[minimal_i])-1:
            alive.remove(minimal_i)
    axis.plot([1-specif for specif in specificity_for_avg],sensitivity_for_avg,color='orange',linewidth=2.5,label=f'mean AUC: {np.mean(auc):.3f} Â± {np.std(auc):.3f}')
    axis.plot([0,1],[0,1],linestyle='dashed',color='k')
    axis.set_xlabel("1-specificity (FPR)")
    axis.set_ylabel("sensitivity (TPR)")
    axis.legend(loc="lower right")
    axis.set_title(f"Logistic Regression (C={label})")
    axis.grid()
```

### *Question 4(d)*

```python
# Fit your logistic regression and apply the above
# function to each selection of C. Which is the optimal value?
fig,axes=plt.subplots(ncols=4,nrows=2,figsize=(17,9))
Cs=[1e-2,1e-1,1,1e1,1e2,1e3,1e4,1e5]

# TO DO

# Solution:
models={c:[] for c in Cs}
for i,c in enumerate(Cs):
    for fold_train,fold_test in folds:
        models[c].append(LogisticRegression(0.001,1000000,C=c))
        models[c][-1].fit(fold_train[0],fold_train[1])
    validate(models[c],[f[1][0] for f in folds],[f[1][1] for f in folds],label=c,axis=axes[i//4,i%4])
```

### *Question 4(e)*

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.
    
def confusion_matrix(predictions,labels):
    
    # TO DO

    # Solution:
    cm=np.zeros((len(np.unique(labels)),len(np.unique(labels))))
    for i,pred in enumerate(predictions):
        cm[pred,int(labels[i])]+=1
    return cm

def precision(predictions,labels,pos_class=1):
    
    # TO DO
    
    # Solution:
    cm=confusion_matrix(predictions,labels)
    return cm[pos_class,pos_class]/sum(cm[pos_class])

def recall(predictions,labels,pos_class=1):
        
    # TO DO
    
    # Solution:
    cm=confusion_matrix(predictions,labels)
    return cm[pos_class,pos_class]/sum(cm[:,pos_class])

def f1_score(predictions,labels):
        
    # TO DO
    
    # Solution:
    r=recall(predictions,labels)
    p=precision(predictions,labels)
    return 2*p*r/(p+r)
```

```python
# Select the appropriate threshold:
thresh_vals=np.arange(0.1,0.9,0.05)

# TO DO

# Soluton:
f1s={t:[] for t in thresh_vals}
for thresh in thresh_vals:
    for fold_train,fold_val in folds:
        model=LogisticRegression(0.001,500000,C=100)
        model.fit(fold_train[0],fold_train[1])
        f1s[thresh].append(np.mean(f1_score(model.predict(fold_val[0],thresh),fold_val[1])))
    print(f"threshold={thresh:.3f}: f1={np.mean(f1s[thresh]):.3f}")
```
