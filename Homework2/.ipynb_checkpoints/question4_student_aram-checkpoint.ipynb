```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.
train_X=np.load("train_X.npy")
train_y=np.load("train_y.npy")>0
test_X=np.load("test_X.npy")
test_y=np.load("test_y.npy")>0

#find the class with smaller number of samples
true_class = 1
false_class = 0

#track indices for upsampling later
inds_true = []
inds_false = []
for i in range(len(train_y)):
    if train_y[i] == true_class:
        inds_true.append(i)
    else:
        inds_false.append(i)

if len(inds_true) > len(inds_false): #more true than false
    smaller_class = 0
else: #more false than true
    smaller_class = 1

#####smaller_class is "1"######

###the rest is coded assuming smaller_class is 1
diff_ = len(inds_false) - len(inds_true) #number of samples to augment
inds_to_upsample = random.sample(inds_true,diff_) #sample the relevant indices
upsample_data = train_X[inds_to_upsample] #get the data
upsample_data = np.array(upsample_data) #make into numpy array...
train_X = np.vstack([train_X,upsample_data]) #augment training data
train_y = np.hstack([train_y,[smaller_class]*diff_]) #augment training data

############# 5 fold datasets ##########
#need to shuffle the data then split into 5 equal parts
num_folds = 5
fold_length = int( len(train_X)/num_folds )

inds_all = np.arange(len(train_X)).tolist()
np.random.shuffle(np.array(inds_all)) #done in place...
train_X = train_X[inds_all]
train_y = train_y[inds_all]

####Create 5 folds of data embedded in a list to be accessed later
X_large = []
y_large = []
for f in range(num_folds):
    inds_start = f*fold_length
    inds_end = (f+1)*fold_length
    X_large.append( train_X[inds_start:inds_end]  )
    y_large.append( train_y[inds_start:inds_end]  )
    
def CreateTrainsSplitValSplit(i,X,Y,total_folds):
    #i is the index to be used as validation data
    #the rest are to create the training data for that fold
    
    
    #get validation data
    val_X = X[i]
    val_y = Y[i]
    
    #get split data from the remaining data
    x_ = []
    y_ = []
    #get training data
    for k in range(total_folds):
        if k!=i: #for all indices except the ones used for the validation set
            x_.append(X[k])
            y_.append(Y[k])

    return np.vstack(x_), np.vstack(y_), val_X, val_y
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
# Auxiliary functions:
def compute_loss(X,y,theta,lam):
    term1 = -sum([y_i*np.log(sigmoid(np.dot(theta,X_i)))+(1-y_i)*np.log(1-sigmoid(np.dot(theta,X_i))) for X_i,y_i in zip(X,y)])
    term2 = lam*np.linalg.norm(theta)**2
    return term1 + term2
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        #copied from my homework 1
        theta_temp = self.theta.reshape(X.shape[1])
        t1 = -1*y@X
        t2 = np.reshape(X@theta_temp,(X.shape[0],1))
        t22 = X.T @ sigmoid(t2)
        grad = t1 + t22.T
        #regularization term
        gradlam = 2*self.lam*self.theta 
        return grad + gradlam #just add them!
    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - X_dataz:    the corresponding validaton folds (features);
# - y_dataz:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.

#make functions to compute true/false positive predictions and rates
def true_pos_predictions(predictions,labels):
    num_true_pos_preds = 0
    for i in range(len(predictions)):
        if labels[i] == 1:
            if predictions[i] == labels[i]:
                num_true_pos_preds+=1
    return num_true_pos_preds
def false_pos_predictions(predictions,labels):
    num_false_pos_preds = 0
    for i in range(len(predictions)):
        if labels[i] == 0:
            if predictions[i] != labels[i]:
                num_false_pos_preds+=1
    return num_false_pos_preds
def true_pos_rate(predictions,labels,pos_class=1):        
    num_trues = sum(labels==pos_class)
    true_pos_preds = true_pos_predictions(predictions,labels)
    return true_pos_preds/num_trues
def false_pos_rate(predictions,labels,false_class=0):
    num_falses = sum(labels==false_class)
    false_pos_preds = false_pos_predictions(predictions,labels)
    return false_pos_preds/num_falses


def validate(models,XXs,YYs,Cval,axis,length_cutoffs=500): 
    #added an extra argument 'length_cutoffs' to change how fine the grid will be
    
    cz = np.linspace(0,1,num=length_cutoffs).tolist()
    
    #keep a running tally of true_pos/false_pos_rate for the means
    #for each cutoff level, to be averaged across the number of models later
    roc_true_mean=np.zeros(length_cutoffs)
    roc_false_mean=np.zeros(length_cutoffs)

    for i, model in enumerate(models):
        #for each model, keep running track of true_pos/false_pos rate in a list
        roc_t = []
        roc_f = []
    
        for c in cz: #for each cutoff level
            pred = model.predict(XXs[i],c) #perform predictions with threshold "c"
            tpr_c = true_pos_rate(pred,YYs[i]) #get true positive rate
            fpr_c = false_pos_rate(pred,YYs[i]) #get false positive rate
            #keep a running track
            roc_t.append(tpr_c)
            roc_f.append(fpr_c)
        
        #plot curve for this model
        axis.plot(roc_f,roc_t, label='roc '+str(i))
        
        for p in range(len(roc_t)):
            roc_true_mean[p] += roc_t[p]
            roc_false_mean[p] += roc_f[p]
    
    axis.plot(cz,cz,linestyle='dashed')
    
    #average true_pos/false_pos_rate across the number of models
    roc_true_mean /= len(models)
    roc_false_mean /= len(models)
    axis.plot(roc_false_mean.tolist(),roc_true_mean.tolist(),color='black',label='mean roc')
    axis.legend()
    
    #print mean AUC score
    auc = abs(np.trapz(roc_true_mean,roc_false_mean,dx=1/(length_cutoffs -1)))
    print('Mean AUC for '+ str(Cval) +': ' + str(auc))
```

### *Question 4(d)*

```python
# Fit your logistic regression and apply the above
# function to each selection of C. Which is the optimal value?
fig,axes=plt.subplots(ncols=4,nrows=2,figsize=(17,9))
Cs=[1e-2,1e-1,1,1e1,1e2,1e3,1e4,1e5]

#index tracker for plotting the axis
ind1 = 0
ind2 = 0

#need to aggregate ALL models 8 instances of 5 models; to be used later
agg_modelz = []

for ax, C in enumerate(Cs):
    #keep a running track per C value
    C_models = []
    all_val_splits_X = []
    all_val_splits_Y = []
    
    #across all folds (in this case, 5 folds)
    for i in range(num_folds):
        #get the split training and validation data for this 
        split_X_train, split_y_train, val_split_X, val_split_y = CreateTrainsSplitValSplit(i,X_large,y_large,num_folds)
        
        #train the model 
        model=LogisticRegression(0.0001,2000,C)
        model.fit(split_X_train,split_y_train)
        
        #collect validation data for this C value
        all_val_splits_X.append(val_split_X)
        all_val_splits_Y.append(val_split_y)
        
        #aggregate models for this value of C
        C_models.append(model)
    #validate + make plots for this value of C
    validate(C_models,all_val_splits_X,all_val_splits_Y,C,axes[ind1][ind2])
    
    #update indices for plotting purposes
    ind2 += 1
    if ind2 == int(len(Cs)/2):
        ind1 = 1
        ind2 = 0
        
    #aggregate models for later
    agg_modelz.append(C_models)
```

### *Question 4(e)*

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.

#compute true/false positive predictions
def true_pos_predictions(predictions,labels):
    num_true_pos_preds = 0
    for i in range(len(predictions)):
        if labels[i] == 1:
            if predictions[i] == labels[i]:
                num_true_pos_preds+=1
    return num_true_pos_preds

def false_pos_predictions(predictions,labels):
    num_false_pos_preds = 0
    for i in range(len(predictions)):
        if labels[i] == 0:
            if predictions[i] != labels[i]:
                num_false_pos_preds+=1
    return num_false_pos_preds

#compute true/false negative predictions
def false_neg_predictions(predictions,labels):
    num_false_neg_preds = 0
    for i in range(len(predictions)):
        if labels[i] == 1:
            if predictions[i] != labels[i]:
                num_false_neg_preds+=1
    return num_false_neg_preds

def true_neg_predictions(predictions,labels):
    num_true_neg_preds = 0
    for i in range(len(predictions)):
        if labels[i] == 0:
            if predictions[i] == labels[i]:
                num_true_neg_preds+=1
    return num_true_neg_preds

#compute the confuction matrix using the above functions; just a 2x2 matrix
def confusion_matrix(predictions,labels):
    cm = np.zeros((2,2))
    #pred class vs actual class
    true_pos = true_pos_predictions(predictions,labels)
    true_neg = true_neg_predictions(predictions,labels)
    false_pos = false_pos_predictions(predictions,labels)
    false_neg = false_neg_predictions(predictions,labels)
    cm[0,0] = true_pos
    cm[0,1] = false_neg
    cm[1,0] = false_pos
    cm[1,1] = true_neg
    return cm

def precision(predictions,labels,pos_class=1):
    tp = true_pos_predictions(predictions,labels)
    fp = false_pos_predictions(predictions,labels)
    if tp == 0 and fp == 0: #catch in case they return all zeros, which has happened sometimes
        return 0
    else:
        return tp/(tp+fp)

def recall(predictions,labels,pos_class=1):
    num_trues = sum(labels==pos_class)
    true_pos_preds = true_pos_predictions(predictions,labels)
    return true_pos_preds/num_trues

def f1_score(predictions,labels):
    r = recall(predictions,labels)
    p = precision(predictions,labels)
    f1 = 2*r*p/(r+p)
    return f1
```

```python
#again, need a sequence of thresholds
cutoffs2=np.linspace(0,1,500)
c_avg = 0

#model with highest AUC is model 6, with C = 1000
best_index = 6 ###changes each time!!!
best_model = agg_modelz[6]

for c in cutoffs2:
    #for each cutoff, compute the fz score
    fz = []
    for i in range(len(all_val_splits_X)):
        preds = best_model[i].predict(all_val_splits_X[i],c)
        fz.append(f1_score(preds,all_val_splits_Y[i]))
        
    #compare against the average f1 score for that c value
    if c_avg < np.mean(np.array(fz)):
        c_best = c
        c_avg = np.mean(np.array(fz))

print(c_best)# TO DO
```
