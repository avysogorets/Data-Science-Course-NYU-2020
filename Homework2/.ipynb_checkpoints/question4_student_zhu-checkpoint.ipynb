```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
train_X=np.load("train_X.npy")
train_y=np.load("train_y.npy")>0
test_X=np.load("test_X.npy")
test_y=np.load("test_y.npy")>0
np.random.seed(2020)
```

### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.

# TO DO
weight = Counter(train_y)
difference = np.abs(weight[False] - weight[True])
minor_class = True if weight[True] < weight[False] else False
selection = np.random.choice(np.where(train_y == minor_class)[0], difference)
train_X_upsampled = np.concatenate((train_X, train_X[selection]))
train_y_upsampled = np.concatenate((train_y, train_y[selection]))


indices = np.arange(len(train_y_upsampled))
np.random.shuffle(indices)
k_fold_indices = np.array_split(indices ,5)


cv_fold = []
for i in range(5):
    cv_fold.append((np.concatenate([k_fold_indices[j] for j in range(5) if i != j]), \
     k_fold_indices[i]))
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
def compute_loss(X,y,theta,lam):
    # TO DO
    return  -sum([y_i*np.log(sigmoid(np.dot(theta,X_i)))+ \
                  (1-y_i)*np.log(1-sigmoid(np.dot(theta,X_i))) \
                  for X_i,y_i in zip(X,y)]) + lam * theta.T @ theta
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        # TO DO
        return (-X.T*y + X.T*sigmoid(self.theta@X.T)).sum(axis=1) + 2* self.lam * self.theta
    def fit(self,X,y):
        np.random.seed(0)
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.

def validate(models,XXs,yys,label,axis):
    
    # TO DO
    tprs = []
    fprs = []
    aucs = []
    interps = []
    for i in range(len(models)):
        y_prob = (models[i].proba(XXs[i]))
        tpr, fpr = roc(y_prob,yys[i])
        tprs.append(tpr)
        fprs.append(fpr)
        axis.plot(fpr, tpr)
        interps.append(np.interp(np.linspace(0,1,100), fpr[::-1], tpr[::-1]))
        aucs.append(((fpr[:-1] - fpr[1:]) * tpr[1:]).sum())
    axis.plot(np.linspace(0,1,100), np.array(interps).mean(axis=0))
    axis.legend(tuple(['Fold {}'.format(f) for f in range(5)] + ['Mean']))
    axis.plot(np.linspace(0,1,100), np.linspace(0,1,100),'--')
    axis.set_title('C = %10.0e,Mean AUC = %.4f'% (label, sum(aucs)/len(aucs)))

def roc(prob, label, pos_label=1):
    tpr = []
    fpr = []
    seq = np.argsort(prob)
    label = label[seq]
    prob = prob[seq]
    for i in prob:
        prediction = prob > i
        TP = ((prediction==label) * (prediction==pos_label)).sum()
        FN = ((prediction!=label) * (prediction!=pos_label)).sum()
        FP = ((prediction!=label) * (prediction==pos_label)).sum()
        TN = ((prediction==label) * (prediction!=pos_label)).sum()
        tpr.append(TP/(TP+FN))
        fpr.append(FP/(FP+TN))
    return np.array(tpr),np.array(fpr)
```

### *Question 4(d)*

```python
# Fit your logistic regression and apply the above
# function to each selection of C. Which is the optimal value?
fig,axes=plt.subplots(ncols=4,nrows=2,figsize=(17,9))
Cs=[1e-2,1e-1,1,1e1,1e2,1e3,1e4,1e5]

# TO DO
i = 0
j = 0
for C in Cs:
    models = []
    XXs = []
    yys = []
    for f in range(5):
        model = LogisticRegression(0.001,20000,C)
        XX = train_X_upsampled[cv_fold[f][0]]
        yy = train_y_upsampled[cv_fold[f][0]]
        model.fit(XX,yy)
        XXs.append(train_X_upsampled[cv_fold[f][1]])
        yys.append(train_y_upsampled[cv_fold[f][1]])
        models.append(model)
    validate(models, XXs, yys, C, axes[i,j])
    j += 1
    if j == 4:
        j = 0
        i += 1
```

I choose $C=100$, since it has the greatest mean AUROC.

### *Question 4(e)*

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.
    
def confusion_matrix(predictions,labels):
    confusion_matrix = pd.crosstab(labels, predictions)
    return confusion_matrix

def precision(predictions,labels,pos_class=1):
    
    # TO DO
    TP = ((predictions == pos_class)*(labels == predictions)).sum()
    FP = ((predictions == pos_class)*(labels != predictions)).sum()
    return TP/(TP + FP)

def recall(predictions,labels,pos_class=1):
        
    TP = ((predictions == pos_class)*(labels == predictions)).sum()
    FN = ((predictions != pos_class)*(labels != predictions)).sum()
    return TP/(TP + FN)

def f1_score(predictions,labels):
        
    p = precision(predictions,labels)
    r = recall(predictions,labels)
    return 2*p*r/(r+p)
```

```python
# Select the appropriate threshold:
thresh_vals=np.arange(0.1,0.9,0.05)
C = 1e2
models = []
XXs = []
yys = []
y_probs =[]
for f in range(5):
    model = LogisticRegression(0.001,20000,C)
    XX = train_X_upsampled[cv_fold[f][0]]
    yy = train_y_upsampled[cv_fold[f][0]]
    model.fit(XX,yy)
    XXs.append(train_X_upsampled[cv_fold[f][1]])
    yys.append(train_y_upsampled[cv_fold[f][1]])
    models.append(model)
    y_probs.append(model.proba(train_X_upsampled[cv_fold[f][1]]))
mean_f1 = np.zeros(len(thresh_vals))
for i in range(len(mean_f1)):
    mean_f1[i] = np.array([f1_score(y_probs[f]>thresh_vals[i], yys[f]) for f in range(5)]).mean()
```
