### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.

def upsample(train_X, train_y):
    class0_idx = np.where(train_y == 0)[0]
    class1_idx = np.where(train_y == 1)[0]

    if len(class0_idx) < len(class1_idx):
        upsampled = np.random.choice(class0_idx, size=len(class1_idx), replace=True)
        train_X = np.concatenate((train_X[upsampled, :], train_X[class1_idx, :]), axis=0) 
        train_y = np.concatenate((train_y[upsampled], train_y[class1_idx]))
    elif len(class0_idx) > len(class1_idx):
        upsampled = np.random.choice(class1_idx, size=len(class0_idx), replace=True)
        train_X = np.concatenate((train_X[upsampled, :], train_X[class0_idx, :]), axis=0) 
        train_y = np.concatenate((train_y[upsampled], train_y[class0_idx]))
    return train_X, train_y

train_X, train_y = upsample(train_X, train_y)

def generate_folds(train_X, train_y, K=5):
    N = len(train_y)
    B = N // K

    perm = np.random.permutation(N)

    train_X_folds = []
    train_y_folds = []
    val_X_folds = []
    val_y_folds = []

    for k in range(K):
        rval = N if k == K - 1 else (k + 1) * B
        val_idx = perm[k * B : rval]
        train_idx = np.delete(perm, slice(k * B, rval))
        
        train_X_folds.append(train_X[train_idx])
        train_y_folds.append(train_y[train_idx])

        val_X_folds.append(train_X[val_idx])
        val_y_folds.append(train_y[val_idx])
    
    return train_X_folds, train_y_folds, val_X_folds, val_y_folds
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
def compute_loss(X,y,theta,lam):
    tx = np.squeeze(np.dot(X, theta.T))
    reg = np.squeeze(np.dot(theta, theta.T))
    return np.sum((1. - y) * tx - np.log(sigmoid(tx))) + lam * reg
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        delta = sigmoid(np.dot(X, self.theta.T)) - y.T  ## (n,1)
        grad = np.sum(X * delta, axis=0, keepdims=True) + 2. * self.lam * self.theta  ## (1,d+1)
        return grad
    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(e)*

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.
    
def confusion_matrix(predictions,labels):
    N = len(predictions)
  cmat = np.zeros((2, 2)).astype(np.int)
  for i in range(N):
    cmat[labels[i]][predictions[i]] += 1
  return cmat

def precision(predictions,labels,pos_class=1):
    cmat = confusion_matrix(predictions,labels)
    tp = cmat[1, 1]
    fp = cmat[0, 1]
    return tp / (tp + fp + 1e-4)

def recall(predictions,labels,pos_class=1):
    cmat = confusion_matrix(predictions,labels)
    tp = cmat[1, 1]
    fn = cmat[1, 0]
    return tp / (tp + fn + 1e-4)

def f1_score(predictions,labels):
    prec = precision(predictions, labels)
    rec = recall(predictions, labels)
    f1 = 2 / (1/rec + 1/prec)
    return f1
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.
def validate(models,XXs,yys,label,axis):
    roc_xy = []
    # auc = []
    for m, X, y in zip(models, XXs, yys):
        y = np.array(y).astype(np.int)
        roc_x, roc_y = [], []
        for t in np.arange(0.0, 1.0 + 1e-3, 0.01):
            pred_y = m.predict(X, t)
            cmat = confusion_matrix(pred_y, y)
            tp = cmat[1,1]
            tn = cmat[0,0]
            fp = cmat[0,1]
            fn = cmat[1,0]
            tpr = tp / (tp + fn)
            fpr = fp / (fp + tn)
            roc_x.append(fpr)
            roc_y.append(tpr)
        axis.plot(roc_x, roc_y, alpha=0.4, c='blue')
        
        roc_xy.append(np.array([roc_x, roc_y]))

        # _auc = 0.01 * (np.sum(roc_y[1:-1]) + (roc_y[0] + roc_y[1]))
        # auc.append(metrics.auc(roc_x, roc_y))

    roc_xy = np.mean(np.array(roc_xy), axis=0)
    auc_mean = 0.01 * (np.sum(roc_xy[1][1:-1]) + (roc_xy[1][0] + roc_xy[1][-1]) / 2)  ## Trapezoidal rule

    axis.plot([0,1],[0,1],'--',c='black')
    axis.plot(roc_xy[0], roc_xy[1], 'r--')
    axis.set_title(f'C = {label}, Mean AUC = {auc_mean:.3f}')

    return auc_mean
```

### *Question 4(d)*

```python
# Fit your logistic regression and apply the above
# function to each selection of C. Which is the optimal value?
fig,axes=plt.subplots(ncols=4,nrows=2,figsize=(17,9))
Cs=[1e-2,1e-1,1,1e1,1e2,1e3,1e4,1e5]

train_X_folds, train_y_folds, val_X_folds, val_y_folds = generate_folds(train_X, train_y)
best_C = 1.
best_auc = 0.0

for row in range(2):
    for col in range(4):
        i = row * 4 + col
        C = Cs[i]

        models = []
        for f in range(5):
            model = LogisticRegression(1e-4, 2000, C)
            model.fit(train_X_folds[f], train_y_folds[f])
            models.append(model)

        mean_auc = validate(models, val_X_folds, val_y_folds, C, axes[row, col])
        if mean_auc > best_auc:
            best_auc = mean_auc
            best_C = C
```

```python
# Select the appropriate threshold:
thresh_vals=np.arange(0.1,0.9,0.05)

f1_all = []
for f in range(5):
    model = LogisticRegression(1e-4, 2000, best_C)
    model.fit(train_X_folds[f], train_y_folds[f])

    f1 = []
    for t in thresh_vals:
        preds = np.array(model.predict(val_X_folds[f], t)).astype(np.int)
        f1.append(f1_score(
            preds,
            np.array(val_y_folds[f]).astype(np.int)
        ))
    
    f1_all.append(f1)

f1_all = np.mean(np.array(f1_all), axis=0)
best_thresh = thresh_vals[np.argmax(f1_all)]
f'The best threshold is {best_thresh}' 
```
