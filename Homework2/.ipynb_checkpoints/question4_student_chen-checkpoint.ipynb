### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
def compute_loss(X,y,theta,lam):
    loss = y @ np.log(sigmoid(X @ theta.T)) + (1-y) @ np.log(1-sigmoid(X @ theta.T))
    loss += lam * np.linalg.norm(theta) ** 2
    return loss
```

```python
# Complete the LogisticRegression class that supports regularization:
class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        grad = -X.T @ (y.T - sigmoid(X @ self.theta.T)) + 2 * self.lam * self.theta.T
        return grad.T
    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.

def auc(x, y):
    coords = list(zip(x, y))
    coords.sort(key = lambda x: x[0]) 
    pairs = list(zip(coords[:-1], coords[1:]))
    return np.sum([ 0.5*(x2-x1)*(y1+y2) for (x1, y1), (x2, y2) in pairs])

def roc(model, X, y, thresholds):
    tpr = np.zeros(len(thresholds))
    fpr = np.zeros(len(thresholds))
    for i, threshold in enumerate(thresholds):
        preds = np.array(model.predict(X, threshold))
        pos_idx = np.argwhere(y == True)
        neg_idx = np.argwhere(y == False)
        num_true_pos = np.sum(preds[pos_idx] == 1)
        num_false_pos = np.sum(preds[neg_idx] == 1)
        tpr[i] = num_true_pos / len(pos_idx)
        fpr[i] = num_false_pos / len(neg_idx)
    return tpr, fpr

def validate(models,XXs,yys,label,axis):
    # TO DO
    thresholds = np.arange(0, 1.01, 0.01)
    tprs = np.zeros((len(models), len(thresholds)))
    fprs = np.zeros((len(models), len(thresholds)))
    aucs = np.zeros(len(models))
    for i, model in enumerate(models):
        tpr, fpr = roc(model, XXs[i], yys[i], thresholds)
        tprs[i,:] = tpr
        fprs[i,:] = fpr
        aucs[i] = auc(fpr, tpr)
        axis.plot(fpr, tpr, label=f"ROC for K={i}")
    axis.plot(np.mean(fprs, axis=0), np.mean(tprs, axis=0), label="Mean ROC")
    axis.plot(thresholds, thresholds, linestyle="dashed")
    axis.legend()
    mean_auc = np.mean(aucs)
    axis.set_title(f"C={label}, mean AUC={mean_auc:.3f}")
```

### *Question 4(e)*

```python
def confusion_matrix(predictions, labels):
    cm = np.zeros((2,2))
    possible_labels = [0, 1]
    for i, predicted_class in enumerate(possible_labels):
        for j, actual_label in enumerate(possible_labels):
            cm[i,j] = sum((predictions == predicted_class) & (labels == actual_label))
    return cm

def precision(predictions,labels,pos_class=1):
    cm = confusion_matrix(predictions, labels)
    return cm[pos_class][pos_class] / np.sum(cm[pos_class,:]) if np.sum(cm[pos_class,:]) > 0 else 0.0

def recall(predictions,labels,pos_class=1):
    cm = confusion_matrix(predictions, labels)
    return cm[pos_class][pos_class] / np.sum(cm[:,pos_class]) if np.sum(cm[:,pos_class]) > 0 else 0.0

def f1_score(predictions,labels):
    p = precision(predictions, labels)
    r = recall(predictions, labels)
    return 2*p*r/(p+r) if p+r>0 else 0.0
```

```python
# Select the appropriate threshold:
thresh_vals=np.arange(0.1,0.9,0.05)

# TO DO
C = 10.0
fs = np.zeros((len(train_X_folds), len(thresh_vals)))
for j, train_X_fold in enumerate(train_X_folds):
    lr = LogisticRegression(alpha=0.001, max_iter=100000, C=C)
    train_y_fold = train_y_folds[j]
    lr.fit(train_X_fold, train_y_fold)
    val_X = np.vstack([ X_fold for k, X_fold in enumerate(train_X_folds) if k != j])
    val_y = np.concatenate([ y_fold for k, y_fold in enumerate(train_y_folds) if k != j]).astype(int)
    for i, thresh in enumerate(thresh_vals):
        preds = np.array(lr.predict(val_X, thresh))
        fs[j, i] = f1_score(preds, val_y)
mean_f1 = np.mean(fs, axis=0)
opt_thresh = thresh_vals[np.argmax(mean_f1)]
print(f"Optimal threshold is {opt_thresh:.2f}, with a mean F1 score of {np.max(mean_f1):.3f}.")
```
