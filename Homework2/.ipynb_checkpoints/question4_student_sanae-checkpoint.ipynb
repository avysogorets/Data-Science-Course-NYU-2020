```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import copy
train_X=np.load("train_X.npy")
train_y=np.load("train_y.npy")>0
test_X=np.load("test_X.npy")
test_y=np.load("test_y.npy")>0
```

### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.

idx_0 = np.where(train_y == 0)[0]
idx_1 = np.where(train_y == 1)[0]

# defining the small and the large classes
big_idx = idx_0 if len(idx_0) > len(idx_1) else idx_1
small_idx = idx_0 if len(idx_0) <= len(idx_1) else idx_1

# upsampling the indexes
upsampled_idx = np.random.choice(small_idx, size=(len(big_idx) - len(small_idx)), replace=True)

# obtaining upsampled data
train_X = np.concatenate((train_X, train_X[upsampled_idx]), axis=0)
train_y = np.concatenate((train_y, train_y[upsampled_idx]), axis=0)


# preparing the k folds
# we shuffle the data since we added all new samples in the end
shuffled_idx = np.arange(len(train_X))
np.random.shuffle(shuffled_idx)
train_X, train_y = train_X[shuffled_idx], train_y[shuffled_idx]

n_fold = 5
fold_len = len(train_X) // 5

train_X_k = np.zeros((n_fold, fold_len, train_X.shape[1]))
train_y_k = np.zeros((n_fold, fold_len))

for i in range(n_fold):
    train_X_k[i] = train_X[i*fold_len:(i+1)*fold_len]
    train_y_k[i] = train_y[i*fold_len:(i+1)*fold_len]
# we throw away one extra data point to keep folds of the same size 
#     if i == n_fold - 1:
#         train_X_k[i] = train_X[i*fold_len:]
#         train_y_k[i] = train_y[i*fold_len:]
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
def compute_loss(X,y,theta,lam):
    # we add the last term to account for regularization
    return -sum([y_i*np.log(sigmoid(np.dot(theta,X_i)))+(1-y_i)*np.log(1-sigmoid(np.dot(theta,X_i))) for X_i,y_i in zip(X,y)]) + lam * np.linalg.norm(theta)**2 
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        # we add the last term to account for regularization
        return - np.transpose(X).dot(y.squeeze() - sigmoid(X.dot(self.theta.squeeze()))) + 2 * self.lam * self.theta.squeeze()

    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.

def validate(models,XXs,yys,label,axis,threshold_num=500):
    # defining the threshold list 
    threshold_list = np.linspace(0,1,threshold_num)
    # average true positive and false positive rate arrays
    av_tpr = np.zeros(threshold_num)
    av_fpr = np.zeros(threshold_num)
    for i, model in enumerate(models):
        tpr = np.array([])
        fpr = np.array([])
        for threshold in threshold_list:
            preds = model.predict(XXs[i], threshold)
            true_pos_rate = 0
            false_pos_rate = 0
            # we coumpote the true positive and false positive rates
            for j, pred in enumerate(preds):
                if pred == 1:
                    if yys[i][j] == 1:
                        true_pos_rate += 1
                    else:
                        false_pos_rate += 1
            true_pos_rate = true_pos_rate / (yys[i]==1).sum()
            false_pos_rate = false_pos_rate/ (yys[i]==0).sum()       
            tpr = np.append(tpr, true_pos_rate)
            fpr = np.append(fpr, false_pos_rate)     
        # plot for this model
        axis.plot(fpr, tpr, label ="model {}".format(i+1))
        av_tpr += tpr
        av_fpr += fpr
    # plot the mean 
    av_tpr = av_tpr / len(models)
    av_fpr = av_fpr / len(models)
    axis.plot(av_fpr, av_tpr, label ="Average ROC")
    axis.plot(threshold_list, threshold_list, linestyle ="dashed")
    axis.legend()
    
    # getting the auc score
    auc = np.abs(np.trapz(av_tpr, av_fpr, dx = 1/(threshold_num-1)))
    
    # printing the auc score
    print("Mean AUC score for C = {} is {}".format(label, auc))
    
    return auc
```

```python
# Fit your logistic regression and apply the above
# function to each selection of C. Which is the optimal value?
fig,axes=plt.subplots(ncols=4,nrows=2,figsize=(17,9))
Cs=[1e-2,1e-1,1,1e1,1e2,1e3,1e4,1e5]
ax1=0
ax2=0
# seemed to be the best learning rate 
lr = 1e-3
best_C = None
best_auc = - np.inf 
for C_value in Cs:
    models = []
    XXs = []
    yys = []
    for k in range(n_fold):
        # for each k we prepate the cross-validation training and validation setes
        valid_X_c, valid_y_c = train_X_k[k], train_y_k[k]
        XXs.append(valid_X_c)
        yys.append(valid_y_c)
        train_X_c = np.concatenate([train_X_k[i] for i in range(n_fold) if i!=k])
        train_y_c = np.concatenate([train_y_k[i] for i in range(n_fold) if i!=k])
        
        # we fit the model 
        model = LogisticRegression(lr, 5000, C_value)
        model.fit(train_X_c, train_y_c)
        
        models.append(model)
        
    auc = validate(models, XXs, yys, C_value, axes[ax1][ax2])
    # storing the best auc score and the C yielding to it 
    if auc > best_auc:
        best_auc = auc
        best_C = C_value
    # keeping track of the axes for the plots
    ax2 += 1
    if ax2 == 4:
        ax2 = 0
        ax1 = 1
print("The value of C yielding the best AUC score is C=", best_C)
```

### *Question 4(e)*

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.

def get_true_positive(preds, labels, pos_class=1):
    # getting the number of true positives 
    true_pos = 0
    for i, pred in enumerate(preds):
        if pred == pos_class and labels[i] == pos_class:
            true_pos += 1
    return true_pos 

def get_false_positive(preds, labels, pos_class=1):
    # getting the number of false positives 
    false_pos = 0
    for i, pred in enumerate(preds):
        if pred == pos_class and labels[i] == 1 - pos_class:
            false_pos += 1
    return false_pos 

def get_true_negative(preds, labels, pos_class=1):
    # getting the number of true negatives 
    true_neg = 0
    for i, pred in enumerate(preds):
        if pred == 1 - pos_class and labels[i] == 1 - pos_class:
            true_neg += 1
    return true_neg 

def get_false_negative(preds, labels, pos_class=1):
    # getting the number of false negatives 
    false_neg = 0
    for i, pred in enumerate(preds):
        if pred == 1 - pos_class and labels[i] == pos_class:
            false_neg += 1
    return false_neg 
    
    
def confusion_matrix(predictions,labels):
    # computing the confusion matrix
    con_matrix = np.zeros((2,2))
    con_matrix[0][0] = get_true_positive(predictions, labels)
    con_matrix[0][1] = get_false_positive(predictions, labels)
    con_matrix[1][0] = get_false_negative(predictions, labels)
    con_matrix[1][1] = get_true_negative(predictions, labels)
    
    return con_matrix
    
    
def precision(predictions,labels,pos_class=1):
    # computing the precision
    true_pos = get_true_positive(predictions, labels, pos_class)
    false_pos = get_false_positive(predictions, labels, pos_class)
    return true_pos / (true_pos + false_pos)

def recall(predictions,labels,pos_class=1):
    # computing the recall
    true_pos = get_true_positive(predictions, labels, pos_class)
    false_neg = get_false_negative(predictions, labels, pos_class)
    return true_pos / (true_pos+false_neg)
    
    
def f1_score(predictions,labels):
    # computing the f1 score 
    precision_value = precision(predictions,labels)
    recall_value = recall(predictions,labels)
    
    return 2 * precision_value * recall_value / (precision_value + recall_value)
    
```

```python
# Select the appropriate threshold:
thresh_vals=np.arange(0.1,0.9,0.05)

# keeping track of the best threshold leading to the best mean f1 score
best_thresh = None
best_f1_score = - np.inf
# iterating over all threshold values 
for thresh in thresh_vals:
    mean_f1_score = 0
    for k in range(n_fold):
        # obtaining the train and validation data
        valid_X_c, valid_y_c = train_X_k[k], train_y_k[k]
        train_X_c = np.concatenate([train_X_k[i] for i in range(n_fold) if i!=k])
        train_y_c = np.concatenate([train_y_k[i] for i in range(n_fold) if i!=k])
        # fitting the model
        model = LogisticRegression(lr, 5000, best_C)
        model.fit(train_X_c, train_y_c)
        preds = model.predict(valid_X_c, thresh)
        # computing the mean f1 score
        mean_f1_score += f1_score(preds, valid_y_c)
        
    mean_f1_score = mean_f1_score/n_fold
    if mean_f1_score > best_f1_score:
        best_f1_score = mean_f1_score
        best_thresh = thresh
```
