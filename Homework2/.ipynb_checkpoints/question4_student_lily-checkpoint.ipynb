### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.
import math


# TO DO
def upsample_minority_class(train_X, train_y):
    classes, counts = np.unique(train_y, return_counts=True)
    minority_class = classes[np.argmin(counts)]
    minority_class_indices, = np.where(train_y == minority_class)
    majority_class_indices, = np.where(train_y != minority_class)
    new_minority_indices = np.random.choice(minority_class_indices, size=max(counts))
    permuted_indices = np.random.choice(
        majority_class_indices.tolist() + new_minority_indices.tolist(),
        size=len(majority_class_indices) + len(new_minority_indices),
        replace=False)
    new_train_X = train_X[permuted_indices]
    new_train_y = train_y[permuted_indices]
    return new_train_X, new_train_y
    
def create_cv_folds(train_X, train_y):
    val_size = len(train_X)/5
    val_indices = [range(int(math.ceil(i*val_size)), int(math.ceil((i+1)*val_size))) for i in range(5)]
    Xtrains = [train_X[~np.array(val_idx)] for val_idx in val_indices]
    Xvals = [train_X[np.array(val_idx)] for val_idx in val_indices]
    ytrains = [train_y[~np.array(val_idx)] for val_idx in val_indices]
    yvals = [train_y[np.array(val_idx)] for val_idx in val_indices]
    return Xtrains, Xvals, ytrains, yvals
    

print(train_X.shape, train_y.shape)
upsampled_train_X, upsampled_train_y = upsample_minority_class(train_X, train_y)
print(upsampled_train_X.shape, upsampled_train_y.shape)
Xtrains, Xvals, ytrains, yvals = create_cv_folds(upsampled_train_X, upsampled_train_y)
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
# Note that I scale the regularization term by n
def compute_loss(X,y,theta,lam):
    
    # TO DO
    loss = -sum([y_i*np.log(sigmoid(np.dot(theta,X_i)))+(1-y_i)*np.log(1-sigmoid(np.dot(theta,X_i))) for X_i,y_i in zip(X,y)]) + len(y) * lam * np.linalg.norm(theta)**2
    return loss
    
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        
        # TO DO
        term_1=sigmoid(np.matmul(self.theta,np.transpose(X)))
        term_2=np.matmul(y,X)
        regularizer = len(y) * self.lam * 2 * self.theta
        return np.matmul(term_1,X)-term_2+regularizer
        
        pass
    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        np.random.seed(3)  # added for consistency of results each time
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.

def validate(models,XXs,yys,label,axis):
    # TO DO
    all_tprs = []
    all_fprs = []
    aucs = []
    for model, X, y in zip(models, XXs, yys):
        tprs = [1.0]
        fprs = [1.0]
        evenly_spaced_tprs = [1.0]
        evenly_spaced_fprs = [1.0]
        preds = np.array(model.proba(X))
        thresholds = sorted(list(set(preds)))
        for thresh in thresholds:
            tp = sum((preds>thresh)&y)
            fp = sum((preds>thresh)&(~y))
            fn = sum((preds<=thresh)&y)
            tprs.append(tp/(tp + fn))
            fprs.append(fp/(fp + fn))
        axis.plot(fprs, tprs, label=label)
        auc = 0
        for tpr1, tpr2, fpr1, fpr2 in zip(tprs[:-1], tprs[1:], fprs[:-1], fprs[1:]):
            auc += np.mean([tpr1, tpr2]) * (fpr1 - fpr2)
        aucs.append(auc)
        
        # plot pointwise average 
        for thresh in np.arange(0, 1, .025):
            tp = sum((preds>thresh)&y)
            fp = sum((preds>thresh)&(~y))
            fn = sum((preds<=thresh)&y)
            evenly_spaced_tprs.append(tp/(tp + fn))
            evenly_spaced_fprs.append(fp/(fp + fn))
        all_tprs.append(evenly_spaced_tprs)
        all_fprs.append(evenly_spaced_fprs)


    # calculate AUC
    auc = 0
    mean_fprs = np.array(all_fprs).mean(axis=0)
    mean_tprs = np.array(all_tprs).mean(axis=0)
    axis.plot(mean_fprs, mean_tprs, label=label, linewidth=4)
    axis.set_title(f'AUC: {np.mean(aucs)}')
    axis.plot(list(np.arange(0, 1.1, .1)), list(np.arange(0, 1.1, .1)), linestyle='--')
```

```python
# Fit your logistic regression and apply the above
# function to each selection of C. Which is the optimal value?
fig,axes=plt.subplots(ncols=4,nrows=2,figsize=(17,9))
Cs=[1e-2,1e-1,1,1e1,1e2,1e3,1e4,1e5]

# TO DO
alpha=.001
max_iter = 20000
for i, C in enumerate(Cs):
    models = []
    for X, y, in zip(Xtrains, ytrains):
        model = LogisticRegression(alpha, max_iter, C)
        model.fit(X, y)
        models.append(model)
    ax = axes[i//4, i%4]
    validate(models, Xvals, yvals, C, ax)
```

The ideal C seems to be 1e3.

### *Question 4(e)*

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.
    
def confusion_matrix(predictions,labels):
    
    # TO DO
    return np.array([
        [sum(~predictions&~labels), sum(~predictions&labels)],
        [sum(predictions&~labels), sum(predictions&labels)],
    ])

def precision(predictions,labels,pos_class=1):
    
    # TO DO
    return sum((predictions==pos_class)&(labels==pos_class))/sum(predictions==pos_class)

def recall(predictions,labels,pos_class=1):
        
    # TO DO
    return sum((predictions==pos_class)&(labels==pos_class))/sum(labels==pos_class)

def f1_score(predictions,labels):
        
    # TO DO
    p = precision(predictions, labels)
    r = recall(predictions, labels)
    return 2 * p * r / (p + r)
```

```python
# Select the appropriate threshold:
thresh_vals=np.arange(0.1,0.9,0.05)

# TO DO
f1s = []
for thresh in thresh_vals:
    thresh_f1s = []
    for Xtrain, ytrain, Xval, yval in zip(Xtrains, ytrains, Xvals, yvals):
        model = LogisticRegression(alpha, max_iter, C=1e2)
        model.fit(Xtrain, ytrain)
        preds = np.array(model.predict(Xval, thresh))
        thresh_f1s.append(f1_score(preds, yval))
    f1s.append(np.mean(thresh_f1s))
plt.plot(thresh_vals, f1s)
f1_dict = dict(zip(thresh_vals, f1s))
print(max(f1_dict, key=f1_dict.get))
```

We get the highest f1 at a threshold of .4

### Question 4(g)

Some disadvantages of the current pipeline and possible improvements:
1. Only using within-season features -> assuming teams stay reasonably similar across seasons, taking into account more information across the seasons could help.
2. The current pipeline throws away rows where there were not enough previous matches to create a window of five to average the statistics -> instead of throwing those rows away, we could keep the rows with a smaller window. If we are concerned about the variances associated with these averages and wish for them to have less of an effect on the learned model, we could consider upsampling the rows from the matches later in the season whose statistics are based on more previous matches.
