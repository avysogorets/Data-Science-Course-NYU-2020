```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
train_X=np.load("train_X.npy")
train_y=np.load("train_y.npy")>0
test_X=np.load("test_X.npy")
test_y=np.load("test_y.npy")>0
```

### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.

# TO DO
class_size_t = train_y.sum()
class_size_f = train_y.shape[0]-train_y.sum()
minority = True if class_size_t<class_size_f else False
balanced_ind = np.random.choice(np.where(train_y==minority)[0], size=max(class_size_t,class_size_f), replace=True)
balanced_X = np.vstack((train_X[balanced_ind],train_X[np.where(train_y!=minority)[0]]))
balanced_y = np.hstack((train_y[balanced_ind],train_y[np.where(train_y!=minority)[0]]))


def split_cv(X,k):
    indices = np.random.randint(0,k, size=X.shape[0])
    return indices

k=5
cv = split_cv(balanced_X,k)
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
def compute_loss(X,y,theta,lam):
    
    # TO DO
    old = -sum([y_i*np.log(sigmoid(np.dot(theta,X_i)))+(1-y_i)*np.log(1-sigmoid(np.dot(theta,X_i))) for X_i,y_i in zip(X,y)])
    return old+lam*np.sum(np.power(theta,2))
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y,theta,lam):
        # TO DO
        old = -np.sum(np.matmul(np.diag(y[0]),X)-np.matmul(np.diag(sigmoid(np.matmul(theta,X.T))[0]),X),axis=0)
        return old+2*lam*theta

    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y,self.theta,self.lam)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.
def cal_rate(prob, true):
    thres = sorted(prob,reverse=True)
    tpr = []
    fpr = []
    #print(prob, true)
    for thr in thres:
        TP = ((prob>=thr)*true).sum()
        FP = ((prob>=thr)*(1-true)).sum()
        FN = ((prob<thr)*true).sum()
        TN = ((prob<thr)*(1-true)).sum()
        #print(TP,FN,FP,TN)
        tpr.append(float(TP) / float(TP+FN))
        fpr.append(float(FP) / float(FP+TN))
    return tpr,fpr




def validate(models,XXs,yys,label,axis):
    # TO DO
    auc = []
    for i in range(len(label)):
        model_C_i = models[i::len(label)]
        axis[i//4][i%4].plot([0,1],[0,1],'--')
        aucs = []
        rocs = {}
        for m in range(len(model_C_i)):
            pred_y_m = model_C_i[m].proba(XXs[m])
            tpr,fpr = cal_rate(pred_y_m,yys[m].astype('int8'))
            rocs[m] = (tpr,fpr)
            AUC = np.trapz(tpr,fpr)
            aucs.append(AUC)
            axis[i//4][i%4].plot(fpr,tpr, label='auc='+str(format(AUC, '.4f')),alpha=0.5)
        auc.append(np.mean(aucs))
        all_fpr = np.unique(np.concatenate([rocs[m][1] for m in rocs.keys()]))
        mean_tpr = np.zeros_like(all_fpr)
        for j in range(len(rocs.keys())):
            mean_tpr += np.interp(all_fpr, rocs[j][1], rocs[j][0])
        mean_tpr /= len(rocs.keys())
        axis[i//4][i%4].plot(all_fpr,mean_tpr, label='pointwise-average roc',color = 'coral',linewidth=3)
        axis[i//4][i%4].legend()
        axis[i//4][i%4].title.set_text('C='+str(label[i])+', mean auc='+str(format(np.mean(aucs), '.4f')))
    plt.show()
    return label[np.argmax(auc)]
    
```

### *Question 4(e)*

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.
    
def confusion_matrix(predictions,labels):
    
    # TO DO
    #suppose the threshold is 0.5
    thr=0.5
    TP = ((predictions>=thr)*labels).sum()
    FP = ((predictions>=thr)*(1-labels)).sum()
    FN = ((predictions<thr)*labels).sum()
    TN = ((predictions<thr)*(1-labels)).sum()
    return [[TN, FP], [FN, TP]]


def precision(predictions,labels,pos_class=1):
    
    # TO DO
    #suppose the threshold is 0.5
    thr=0.5
    if pos_class==0:
        labels = 1-labels
    TP = ((predictions>=thr)*labels).sum()
    FP = ((predictions>=thr)*(1-labels)).sum()
    return float(TP)/(TP+FP)

def recall(predictions,labels,pos_class=1):
        
    # TO DO
    #suppose the threshold is 0.5
    thr=0.5
    if pos_class==0:
        labels = 1-labels
    TP = ((predictions>=thr)*labels).sum()
    FN = ((predictions<thr)*labels).sum()
    return float(TP)/(TP+FN)

def f1_score(predictions,labels,thr):
        
    # TO DO
    #suppose positive class =1
    TP = ((predictions>=thr)*labels).sum()
    FN = ((predictions<thr)*labels).sum()
    FP = ((predictions>=thr)*(1-labels)).sum()
    p = float(TP)/(TP+FP)
    r = float(TP)/(TP+FN)
    return 2*p*r/(p+r)
    
    
```

```python
# Select the appropriate threshold:
thresh_vals=np.arange(0.1,0.9,0.05)

# TO DO
C = validate(models,XXs,yys,Cs,axes)
model_C = models[Cs.index(C)::len(Cs)]
f1 = []
for i in range(len(thresh_vals)):
    f1s = []
    for m in range(len(model_C)):
        pred_y_m = model_C[m].proba(XXs[m])
        F1 = f1_score(pred_y_m,yys[m].astype('int8'),thresh_vals[i])
        f1s.append(F1)
    f1.append(np.mean(f1s))
    
thresh_vals[np.nanargmax(f1)]
 
```
