### *Question 4(a)*

```python
# Upsample the smaller class in the training data to ensure perfect
# class balance. Then split it into 5 folds for cross-validation.
# You should not use sklearn.
def upsample(train_X, train_y):
    y1_percent = train_y.mean()
    smaller_target = 0
    if y1_percent < 0.5:
        smaller_target = 1
    smaller_X, smaller_y = train_X[train_y == smaller_target], train_y[train_y == smaller_target]
    larger_X, larger_y = train_X[~(train_y == smaller_target)], train_y[~(train_y == smaller_target)]
    idx = np.random.randint(0,len(smaller_X),len(larger_X))
    b_smaller_X, b_smaller_y = smaller_X[idx], smaller_y[idx]
    X = np.concatenate([b_smaller_X, larger_X])
    y = np.concatenate([b_smaller_y, larger_y])
    idx = list(range(len(X)))
    np.random.shuffle(idx)

    X = X[idx]
    y = y[idx]
    return X, y


def mykfold(X,y,k):
    idx = list(range(len(X)))
    np.random.shuffle(idx)
    idx = np.array(idx)
    idxs = [idx[len(idx)//k * i: len(idx)//k * (i+1)] for i in range(k-1)]
    idxs.append(idx[len(idx)//k * (k-1):])    
    
    for i in range(k):
        idx_idxs = list(range(k))
        idx_idxs.remove(i)
        train_idx = idx[np.concatenate([idxs[j] for j in idx_idxs])]
        test_idx  = idx[idxs[i]]
        yield train_idx, test_idx

train_X, train_y = upsample(train_X, train_y)
folds = [(train_X[train], train_y[train], train_X[valid], train_y[valid]) for (train, valid) 
 in mykfold(train_X, train_y, 5)]
# TO DO
```

### *Question 4(b)*

```python
# Copied from Homework 1:
def sigmoid(t):
    return 1./(1+np.exp(-t))
```

```python
# Add the L2 regularization term to the old loss function: 
def compute_loss(X,y,theta,lam):
    
    # TO DO
    return -sum([y_i*np.log(sigmoid(np.dot(theta,X_i)))+(1-y_i) * 
                 np.log(1-sigmoid(np.dot(theta,X_i))) for X_i,y_i in zip(X,y)]) + lam* (theta.flatten()**2).sum()
```

```python
# Complete the LogisticRegression class that supports regularization:

class LogisticRegression():
    def __init__(self,alpha,max_iter,C,tolerance=1e-5):
        self.lam=1/C
        self.alpha=alpha
        self.tolerance=tolerance
        self.max_iter=max_iter
    def gradient(self,X,y):
        
        # TO DO
        t = np.matmul(X,self.theta.reshape((self.theta.shape[1],1)))
    
        st = sigmoid(t)

        y_st = y.reshape((y.shape[1],1)) - st
   
        X_y_st = np.multiply(X, y_st)
        
        return -X_y_st.sum(axis=0).reshape(self.theta.shape) + 2 * self.lam * self.theta
    
        
    def fit(self,X,y):
        y=y.reshape((1,X.shape[0]))
        X=np.hstack([X,np.ones((X.shape[0],1))])
        self.theta=np.random.normal(loc=0,scale=1,size=X.shape[1]).reshape((1,X.shape[1]))
        progress=[0,compute_loss(X,y[0],self.theta[0],self.lam)]
        while self.max_iter>=0 and abs(progress[-1]-progress[-2])>self.tolerance:
            self.theta-=self.alpha*self.gradient(X,y)
            if self.max_iter%500==0:
                progress.append(compute_loss(X,y[0],self.theta[0],self.lam))
            self.max_iter-=1
        return progress[1:]
    def margin(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.matmul(self.theta,np.transpose(X)))
    def predict(self,X,thresh):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return [1 if sigmoid(np.dot(self.theta,X_i))>thresh else 0 for X_i in X]
    def proba(self,X):
        X=np.hstack([X,np.ones((X.shape[0],1))])
        return np.squeeze(np.array([sigmoid(np.dot(self.theta,X_i)) for X_i in X]))
```

### *Question 4(c)*

```python
# Complete the below function. Arguments:
# - models: K models from the K-fold cross-validation;
# - XXs:    the corresponding validaton folds (features);
# - yys:    the corresponding validaton folds (targets);
# - label:  regularization constant (C);
# - axis:   a subplot to be used for plotting;

# The function should plot the ROC curve for each of the
# models as well as their pointwise average (highlighted).
# Moreover, plot the diagonal (0,0)-(1,1) as a dashed line
# and print the mean AUC score.

def myroc(y_pred, y, threshold=None):
    y_pred_y = np.concatenate([y_pred[:, np.newaxis], y[:, np.newaxis]], axis=1)
    y_pred_unique = np.unique(y_pred)
    if threshold is not None:
        y_pred_unique = threshold
    np.sort(y_pred_unique)
    
    tp, fp = [],[]
    for i, thre in enumerate(y_pred_unique):
        
        thre_tp = (y_pred_y[y_pred_y[:,0]>= thre][:,1]==1).sum()
        thre_fp =  (y_pred_y[y_pred_y[:,0]>= thre][:,1] == 0).sum()
        tp.append(thre_tp)
        fp.append(thre_fp)
    tp.append(0)
    fp.append(0)
    
    tpr, fpr = np.flip( np.array(tp) / (np.array(y)==1).sum()), np.flip(np.array(fp) / (np.array(y)==0).sum())
    
    return  tpr, fpr

def myauc(tpr,fpr):
    if tpr[1] == fpr[1] == 1:
        return 0.5
    area = 0
    for i in range(len(tpr)-1):
        area += (fpr[i+1] - fpr[i])*tpr[i] 
    return area
        
def validate(models,XXs,yys,label,axis):
    plt.subplot(2,4,axis)
    interp = []
    auc = 0
    for i in range(len(models)):
        pred_y = models[i].proba(XXs[i])
        tpr, fpr = myroc(pred_y, yys[i])
        auc += myauc(tpr, fpr)
        
        if tpr[1] == fpr[1] == 1:
            interp_iter = np.linspace(0,1,100)
            interp.append(interp_iter)
        else:
            interp_iter = [0]
            for j in np.linspace(0,1,100)[1:]:
                interp_iter.append(tpr[fpr <=j][-1])
            interp.append(interp_iter)

        plt.plot(fpr,tpr)
    plt.plot(np.linspace(0,1,100) ,np.array(interp).mean(axis=0), label='average')
    plt.plot([0,1], [0,1], '--')  
    plt.title('Average AUC %.3f' % (auc/len(models)))
    plt.legend()
    
```

```python
# Complete the below functions. The returned confusion matrix
# should be a two-dimensional NumPy array.
    
def confusion_matrix(predictions,labels):
    tp = predictions[labels == 1].sum() 
    tn = (predictions[labels == 0] == 0).sum()
    
    fp =  (predictions[labels == 0] == 1).sum() 
    fn = (predictions[labels == 1] == 0).sum() 
    
    return [[tp,fn],[fp,tn]]
    # TO DO

    
def precision(predictions,labels,pos_class=1):
    return labels[predictions == pos_class].sum() / float( (predictions == pos_class).sum())
    
    # TO DO


def recall(predictions,labels,pos_class=1):
    return predictions[labels == pos_class].sum() / float(( labels == pos_class).sum())
    # TO DO
    


def f1_score(predictions,labels):
    p = precision(predictions, labels) 
    r = recall(predictions, labels)
    return 2* p*r / (p+r)

    # TO DO
    
```
