```python
import os
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score

train=pd.read_csv('train_elections.csv',index_col=0)
test=pd.read_csv('test_elections.csv',index_col=0)
features=train.columns[:-2]
target="trumpWinner"
```

```python
# Part (a):

import graphviz
import pydotplus
from sklearn.tree import export_graphviz as export
from sklearn.tree import DecisionTreeClassifier as DTC

def visualize(skl_tree):
    dot_format=export(skl_tree,class_names=["Clinton","Trump"],filled=True,feature_names=features)
    pydot_graph=pydotplus.graph_from_dot_data(dot_format)
    pydot_graph.set_size('"4,4!"')
    display(graphviz.Source(pydot_graph.to_string()))

'''Fit a decision tree on your train data and visualize it'''
target="trumpWinner"
model = DTC(criterion='entropy',min_samples_split=5)
model.fit(train.loc[:,features],train.loc[:,target])
visualize(model)

print(features)
```

```python
'''
Recreate this tree using if-else statements by completing the below function.
 - argument: a Pandas DataFrame corresponding to a single test state;
 - return: a boolean prediction for the target variable;
'''

def tree_alias(inp):
    # TO DO
    if inp['trump'] <= 74.5:
        if inp['gun'] <= 80.5:
            return 1
        else:
            if inp['obamacare'] <= 91.0:
                return 1
            else:
                return 0
    else:
        if inp['gun'] <= 63:
            return 1
        else:
            return 0
        
    return None

```

```python
# Check your results (run this cell):
ifelse_preds=[tree_alias(test.loc[state]) for state in test.index]
sklearn_preds=model.predict(test.loc[:,features])
results=[[int(ie),int(sk),int(not ie^sk)] for ie,sk in zip(ifelse_preds,sklearn_preds)]
results_df=pd.DataFrame(results,columns=['tree alias','tree sklearn','match?'],index=test.index)
display(results_df.T)
```

```python
# Part (b):

import math
def extended_log(arg):
    return 0 if not arg else math.log(arg)
def entropy(p):
    return -sum([p_i*extended_log(p_i) for p_i in p])

'''
Complete the class 'decision stump' that has attributes/methods:
- attribute 'self.split_feature': the name of the optimal splitting feature;
- attribute 'self.split_value': value at which self.split_feature should be split on;
- method 'fit(self,X,y)': decides on the above two attributes based on given Pandas
DataFrames X and y holding features and targets for all training examples, respectively.
Returns the list [self.split_feature,self.split_value] after fitting.
- method 'predict(self,X)': returns a list of predictions given a Pandas DataFrame X
holding features of some number of test examples. 
'''

class decision_stump():

    def __init__(self):
        self.split_feature=None
        self.split_value=None
        
    def fit(self,X,y):
        # TO DO
       
        prob_y = np.asarray([(y==True).sum() / len(y), (y==False).sum() / len(y) ])
        entropy_y = entropy(prob_y)
        entropy_feat = []
        vmin_feat = []
         

        for feat in features:
            X_i = X.loc[:,feat]
            setX = set(X_i.tolist())
            rmin, rmax = min(setX), max(setX)
            
            vmin = 0
            entropy_min = np.inf
            #for setx_i in setX:
            for setx_i in range(rmin, rmax):
                pred0, pred1 = (X_i < setx_i), (X_i >= setx_i)
                y_1,y_0 = y[pred1], y[pred0] 
                p_xy1 = sum(y_1) / len(y_1) if len(y_1) >= 1 else 0 
                p_xy0 = sum(y_0) / len(y_0) if len(y_0) >= 1 else 0
                prob_xy = np.asarray([p_xy1, p_xy0])

                PXY1 = np.asarray([1-p_xy1, p_xy1])
                PXY0 = np.asarray([1-p_xy0,p_xy0])
                p_x0  = np.sum(pred0)/len(pred0)
                p_x1 = np.sum(pred1)/len(pred1)
                entropy_i = p_x1*entropy(PXY1) +  p_x0 *entropy(PXY0)

                if entropy_min > entropy_i:
                    entropy_min = entropy_i
                    vmin = setx_i

            entropy_feat.append(entropy_y-entropy_min)
            #entropy_feat.append(entropy_max)
            vmin_feat.append(vmin)

        entropy_feat = np.asarray(entropy_feat)
        maxind = np.argmax(entropy_feat)
        
        self.split_feature = features[maxind]
        self.split_value = vmin_feat[maxind]
        return [self.split_feature, self.split_value]
    
    def predict(self,X):
        # TO DO
        return self.model.predict(X)
               
    
    def predict(self,X):
        # TO DO
        return self.model.predict(X)
```

```python
# Part (c):
from sklearn.ensemble import RandomForestClassifier as RFC
'''
Fit a random forest of 200 sklearn's decision stumps. Use entropy as splitting
criterion, bootstrap sampling of the training data and consider a random subset
of only 5 features for each split (effectively, for each tree). 
'''

# TO DO
num_learners=200
model_2=RFC(n_estimators=num_learners,criterion='entropy',bootstrap=True,max_features=5, max_depth=1)
model_2.fit(train.loc[:,features],train.loc[:,target])

# We can extract individual trees:
print(f'we fitted {len(model_2.estimators_)} decision trees!')

def parse(label):
    clinton=int(label.split('nvalue = [')[1].split(', ')[0])
    trump=int(label.split(']\\nclass')[0].split(', ')[1])
    return  [clinton,trump]

# Visualize trees one by one:
tree=model_2.estimators_[39] # change this!
dot_format=export(tree,class_names=["Clinton","Trump"],filled=True,feature_names=features)
pydot_graph=pydotplus.graph_from_dot_data(dot_format)
pydot_graph.set_size('"4,4!"')
for node in pydot_graph.get_node_list():
    if node.get_label(): # the first node is 'None' 
        clinton,trump=parse(node.get_label())
        if clinton>trump:
            color='#{a:02x}{a:02x}ff'.format(a=int(2*255*trump/(trump+clinton)))
        else:
            color='#ff{a:02x}{a:02x}'.format(a=int(2*255*clinton/(trump+clinton)))
        node.set_fillcolor(color)
graphviz.Source(pydot_graph.to_string())
```
