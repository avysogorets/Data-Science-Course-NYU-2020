```python
# Part (b):

import math
def extended_log(arg):
    return 0 if not arg else math.log(arg)
def entropy(p):
    return -sum([p_i*extended_log(p_i) for p_i in p])

'''
Complete the class 'decision stump' that has attributes/methods:
- attribute 'self.split_feature': the name of the optimal splitting feature;
- attribute 'self.split_value': value at which self.split_feature should be split on;
- method 'fit(self,X,y)': decides on the above two attributes based on given Pandas
DataFrames X and y holding features and targets for all training examples, respectively.
Returns the list [self.split_feature,self.split_value] after fitting.
- method 'predict(self,X)': returns a list of predictions given a Pandas DataFrame X
holding features of some number of test examples. 
'''

class decision_stump():
    def __init__(self):
        self.split_feature=None
        self.split_value=None
        self.left_label=None
        self.right_label=None

    def __ig(self, X, y, split_feature, split_value):
        left = X[X[split_feature] <= split_value]
        left_y = y[X[split_feature] <= split_value]
        right = X[X[split_feature] > split_value]
        right_y = y[X[split_feature] > split_value]
        n = len(X)
        X_p = y.value_counts() / n
        left_p = left_y.value_counts() / len(left_y)
        right_p = right_y.value_counts() / len(right_y)
        return entropy(X_p) - (len(left) * entropy(left_p) + len(right) * entropy(right_p)) / n
        
    def fit(self,X,y):
        max_ig = None
        argmax_split_feature = None
        argmax_split_value = None
        for feature in X.columns:
            for val in X[feature]:
                ig = self.__ig(X, y, feature, val)
                if max_ig is None or ig > max_ig:
                    max_ig = ig
                    argmax_split_feature = feature
                    argmax_split_value = val
        self.split_feature = argmax_split_feature
        self.split_value = argmax_split_value
        left_y = y[X[self.split_feature] <= self.split_value]
        right_y = y[X[self.split_feature] > self.split_value]
        self.left_label = left_y.value_counts(ascending=False).axes[0][0]
        self.right_label = right_y.value_counts(ascending=False).axes[0][0]
        return [self.split_feature, self.split_value]
    
    def predict(self,X):
        return X.apply(lambda row: self.left_label if row[self.split_feature] <= split_value else self.right_label, axis=1)
```

```python
'''
Compare your implementation to one from sklearn; in particular, fit a sklearn's
decision stump on the same training data, visualize the resulting tree and compare
to your model in terms of the splitting feature and the splitting value. Comment
on your findings.
'''
ds = decision_stump()
split_feature, split_value = ds.fit(train.loc[:,features], train.loc[:,target])
print(f'Split feature: {split_feature}, split_value: {split_value:.0f}')

sklearn_stump = DTC(max_depth=1, min_samples_leaf=1)
sklearn_stump.fit(train.loc[:,features],train.loc[:,target])
visualize(sklearn_stump)
```

I obtained the same split feature as sklearn's decision tree stump, but got a split value that was 0.5 off. However, this still results in the same split of the training data as the sklearn clasisfier. Since the value of this feature is always an integer, it will also always result in the same split of test data. However, if this feature had been a float, it might have been better to choose a split value that was interpolated between the max value of the data in the left child node and the min value of the data in the right child node.

```python
# Part (c):
from sklearn.ensemble import RandomForestClassifier as RFC
'''
Fit a random forest of 200 sklearn's decision stumps. Use entropy as splitting
criterion, bootstrap sampling of the training data and consider a random subset
of only 5 features for each split (effectively, for each tree). 
'''

# TO DO
num_learners=200
model_2=RFC(n_estimators=num_learners,criterion='entropy',bootstrap=True,max_features=5)
model_2.fit(train.loc[:,features],train.loc[:,target])
```

```python
'''
Add the 'random forest correct?' row to the table we created in the previous part
How does it perform compared to the decision stump? Compared to the decision tree?
'''

# TO DO
rfc_preds = model_2.predict(test.loc[:,features])
results = [[int(label),int(not sk_pred ^ label),int(not ds_pred ^ label),int(not rfc_pred ^ label)] for sk_pred, label, ds_pred, rfc_pred in zip(sklearn_preds, ground_truth, ds_preds, rfc_preds) ]
results_df = pd.DataFrame(results, columns = ['truth', 'decision tree correct?', 'decision stump correct?', 'random forest correct?'],index=test.index)
display(results_df.T)
ds_correct=np.sum([not p^t for p,t in zip(ds_preds, test.loc[:,target])])
dt_correct=np.sum([not p^t for p,t in zip(sklearn_preds, test.loc[:,target])])
rfc_correct=np.sum([not p^t for p,t in zip(rfc_preds, test.loc[:,target])])
print(f'correct decision stump test predictions: {ds_correct} out of {len(test)} ({100*ds_correct/len(test):.0f}%)')
print(f'correct decision tree test predictions: {dt_correct} out of {len(test)} ({100*dt_correct/len(test):.0f}%)')
print(f'correct random forest test predictions: {rfc_correct} out of {len(test)} ({100*rfc_correct/len(test):.0f}%)')
```

The random forest classifier beats both the decision stump and decision tree by 8 percentage points.

```python
'''
Create a Pandas Dataframe showing one sample standard deviation of predictions across
all learners in the forest for each test state (hint: we did it in lab 2). Add the 
'random forest correct?' row to the table. Comment on your findings.
'''

# TO DO
# Estimate uncertainty of predictions:
predictions=[]
for tree in model_2.estimators_:
    predictions.append(tree.predict(test.loc[:,features]))
def uncertainty(x,y,y_hat):
    y_hat_x=[prediction[x] for prediction in y_hat]
    sigma_hat=np.sqrt(np.divide(sum([(p-y[x])**2 for p in y_hat_x]),num_learners))
    return sigma_hat
difficulties=pd.DataFrame([[round(uncertainty(obs,test.loc[:,target],predictions),3), int(not rfc_pred ^ label)] for obs, rfc_pred, label in zip(range(len(test)), rfc_preds, ground_truth)],
                          index=test.index,columns=["sigma_hat", "random forest correct?"])
display(difficulties.T)
```

The random forest classifier exhibited the greatest uncertainty (standard deviation) in the states where it incorrectly classified the winner.
