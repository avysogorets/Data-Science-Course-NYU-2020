```python
import numpy as np
import pandas as pd
train=pd.read_csv('train_elections.csv',index_col=0)
test=pd.read_csv('test_elections.csv',index_col=0)
features=train.columns[:-2]
target="trumpWinner"
```

```python
# Part (a):

import graphviz
import pydotplus
from sklearn.tree import export_graphviz as export
from sklearn.tree import DecisionTreeClassifier as DTC

def visualize(skl_tree):
    dot_format=export(skl_tree,class_names=["Clinton","Trump"],filled=True,feature_names=features)
    pydot_graph=pydotplus.graph_from_dot_data(dot_format)
    pydot_graph.set_size('"4,4!"')
    display(graphviz.Source(pydot_graph.to_string()))

'''Fit a decision tree on your train data and visualize it'''

# TO DO
model = DTC(criterion="entropy", min_samples_split=5, random_state=1)
model.fit(train[features], train[target])
visualize(model)
```

```python
'''
Recreate this tree using if-else statements by completing the below function.
 - argument: a Pandas DataFrame corresponding to a single test state;
 - return: a boolean prediction for the target variable;
'''

def tree_alias(inp):
    if inp['trump'] <= 74.5:
        if inp['gun'] <= 63.0:
            return False
        else:
            return True
    else:
        if inp['gun'] <= 80.5:
            if inp['obamacare'] <= 91.0:
                return False
            else:
                return True
        else:
            return True
    return None
```

```python
# Check your results (run this cell):
ifelse_preds=[tree_alias(test.loc[state]) for state in test.index]
sklearn_preds=model.predict(test.loc[:,features])
results=[[int(ie),int(sk),int(not ie^sk)] for ie,sk in zip(ifelse_preds,sklearn_preds)]
results_df=pd.DataFrame(results,columns=['tree alias','tree sklearn','match?'],index=test.index)
display(results_df.T)
```

```python
# Part (b):

import math
def extended_log(arg):
    return 0 if not arg else math.log(arg)
def entropy(p):
    return -sum([p_i*extended_log(p_i) for p_i in p])

'''
Complete the class 'decision stump' that has attributes/methods:
- attribute 'self.split_feature': the name of the optimal splitting feature;
- attribute 'self.split_value': value at which self.split_feature should be split on;
- method 'fit(self,X,y)': decides on the above two attributes based on given Pandas
DataFrames X and y holding features and targets for all training examples, respectively.
Returns the list [self.split_feature,self.split_value] after fitting.
- method 'predict(self,X)': returns a list of predictions given a Pandas DataFrame X
holding features of some number of test examples. 
'''

class decision_stump():
    def __init__(self):
        self.split_feature=None
        self.split_value=None
    def fit(self,X,y):
        num_samples = len(X)
        curr_best_ig = None
        class_mask = y
        for feature in X.columns:
            min_val = min(X[feature])
            max_val = max(X[feature])
            orig_entropy = entropy([
                sum(class_mask)/num_samples, 1 - sum(class_mask)/num_samples])
            for val in range(min_val, max_val):
                split_mask = X[feature] <= val
                p_lower_leaf = len(X[split_mask & class_mask])/len(X[split_mask])
                p_upper_leaf = len(X[~split_mask & class_mask])/len(X[~split_mask])
                new_entropy = (
                    len(X[split_mask])/num_samples * entropy([p_lower_leaf, 1 - p_lower_leaf]) +
                    len(X[~split_mask])/num_samples * entropy([p_upper_leaf, 1 - p_upper_leaf])
                )
                ig = orig_entropy - new_entropy
                if curr_best_ig is None or ig > curr_best_ig:
                    self.split_feature = feature
                    self.split_value = val
                    curr_best_ig = ig
                    self.curr_best_ig = curr_best_ig
    def predict(self,X):
        return X[self.split_feature] > self.split_value
```

```python
'''
Compare your implementation to one from sklearn; in particular, fit a sklearn's
decision stump on the same training data, visualize the resulting tree and compare
to your model in terms of the splitting feature and the splitting value. Comment
on your findings.
'''

# TO DO
ds = decision_stump()
ds.fit(train[features], train[target])
print(ds.split_feature, ds.split_value)
sklearn_decision_stump = DTC(max_depth=1, criterion="entropy")
sklearn_decision_stump.fit(train[features], train[target])
visualize(sklearn_decision_stump)
```

The results match up to the singles digit (e.g. <=74.5 vs. <=74). They are equivalent for this data though, since the feature values can only take on integer values. 


```python
'''
Create another table like 'results_df' for the test set with the following rows:
(a) truth, (b) decision tree correct?, (c) decision stump correct? Does your decision
stump perform worse than the tree you fitted before? Are mistakes the same or not?
Comment on advantages and disadvatages of a decision stump compared to a decision tree.
'''

# TO DO
preds = ds.predict(test[features])
results_df1 = pd.DataFrame()
results_df1['truth'] = results_df['truth']
results_df1['decision tree correct?'] = results_df['correct?']
results_df1['decision stump correct?'] = preds == test[target]
results_df1
```

The decision stump makes more mistakes than the decision tree, primarily due to underfitting. The mistakes between the two models are different (e.g. the decision tree gets right all the samples the decision stump gets wrong, and vice versa). The primary advantage of a decision stump is its simplicity, i.e. it's just a single if/else rule. It is also very hard to overfit a decision stump, in comparison to a decision tree (especially if the latter is not sufficiently regularized). The primary disadvantage of a decision stump is its lack of expressivity since the model is limited to just one split.

```python
# Part (c):
from sklearn.ensemble import RandomForestClassifier as RFC
'''
Fit a random forest of 200 sklearn's decision stumps. Use entropy as splitting
criterion, bootstrap sampling of the training data and consider a random subset
of only 5 features for each split (effectively, for each tree). 
'''

rfc = RFC(max_depth=1, criterion="entropy", n_estimators=200, max_features=5)
rfc.fit(train[features], train[target])
```

```python
'''
Add the 'random forest correct?' row to the table we created in the previous part
How does it perform compared to the decision stump? Compared to the decision tree?
'''

preds = rfc.predict(test[features])
results_df1['random forest correct?'] = preds == test[target]
results_df1
```

The random forest classifier performs better than a single decision stump but still not as well as a decision tree.

```python
'''
Create a Pandas Dataframe showing one sample standard deviation of predictions across
all learners in the forest for each test state (hint: we did it in lab 2). Add the 
'random forest correct?' row to the table. Comment on your findings.
'''

def uncertainty(x,y,y_hat, num_learners=200):
    y_hat_x=[prediction[x] for prediction in y_hat]
    sigma_hat=np.sqrt(np.divide(sum([(p-y[x])**2 for p in y_hat_x]),num_learners))
    return sigma_hat

predictions=[]
for tree in rfc.estimators_:
    predictions.append(tree.predict(test.loc[:,features]))
    
results_df1['random forest uncertainty'] = [uncertainty(i,test.loc[:, target],predictions) for i in range(len(test))]
```

```python
results_df1
```

The uncertainty measured in the standard deviation of predictions across the decision stumps in the random forest tends to be lower for the predictions the model gets correct. This is a nice result since it shows the model is reasonably calibrated.
