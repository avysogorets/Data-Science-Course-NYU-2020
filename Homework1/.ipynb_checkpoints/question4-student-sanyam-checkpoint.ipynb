```python
import numpy as np
import pandas as pd
train=pd.read_csv('train_elections.csv',index_col=0)
test=pd.read_csv('test_elections.csv',index_col=0)
features=train.columns[:-2]
target="trumpWinner"
```

```python
# Part (a):

import graphviz
import pydotplus
from sklearn.tree import export_graphviz as export
from sklearn.tree import DecisionTreeClassifier as DTC

def visualize(skl_tree):
    dot_format=export(skl_tree,class_names=["Clinton","Trump"],filled=True,feature_names=features)
    pydot_graph=pydotplus.graph_from_dot_data(dot_format)
    #pydot_graph.set_size('"4,4!"')
    display(graphviz.Source(pydot_graph.to_string()))

'''Fit a decision tree on your train data and visualize it'''

model = DTC(criterion='entropy', min_samples_split=5, random_state=43)
model.fit(train.loc[:,features], train.loc[:,target])

visualize(model)
```

```python
'''
Recreate this tree using if-else statements by completing the below function.
 - argument: a Pandas DataFrame corresponding to a single test state;
 - return: a boolean prediction for the target variable;
'''

def tree_alias(inp):
    if inp['trump'] <= 74.5:
        return inp['gun'] > 63.0
    else:
          if inp['gun'] <= 80.5:
            return inp['obamacare'] > 91.0
          else:
            return True
```

```python
# Check your results (run this cell):
ifelse_preds=[tree_alias(test.loc[state]) for state in test.index]
sklearn_preds=model.predict(test.loc[:,features])
results=[[int(ie),int(sk),int(not ie^sk)] for ie,sk in zip(ifelse_preds,sklearn_preds)]
results_df=pd.DataFrame(results,columns=['tree alias','tree sklearn','match?'],index=test.index)
display(results_df.T)
```

```python
'''
Fetch predictions from the fitted model (or use your 'tree_alias function')
and compare them to ground truth on the test set. In particular, create a Pandas
DataFrame named 'results_df' as above but with 'predicted', 'truth', 'correct?'
as rows and test states as columns. How many are classified correctly?
'''

results_df = pd.DataFrame({
  'predicted': model.predict(test.loc[:,features]),
  'truth': test.loc[:,target]
})
results_df['correct?'] = results_df.apply(lambda row: not (row['predicted'] ^ row['truth']), axis=1)
results_df = results_df.T

print(f"correct test predictions: {sum(results_df.apply(lambda row: row['correct?'], axis=0))} out of {len(test)}")

results_df
```

```python
# Part (b):

import math
def extended_log(arg):
    return 0 if not arg else math.log(arg)
def entropy(p):
    return -sum([p_i*extended_log(p_i) for p_i in p])

'''
Complete the class 'decision stump' that has attributes/methods:
- attribute 'self.split_feature': the name of the optimal splitting feature;
- attribute 'self.split_value': value at which self.split_feature should be split on;
- method 'fit(self,X,y)': decides on the above two attributes based on given Pandas
DataFrames X and y holding features and targets for all training examples, respectively.
Returns the list [self.split_feature,self.split_value] after fitting.
- method 'predict(self,X)': returns a list of predictions given a Pandas DataFrame X
holding features of some number of test examples. 
'''

class decision_stump:
    def __init__(self):
        self.split_feature = None
        self.split_value = None
    
    def fit(self,X,y):
        def _compute_binary_entropy(p):
            return - p * np.log(p + 1e-6) - (1 - p) * np.log(1 - p + 1e-6)
      
        def _compute_split_entropy(f, t):
            lsplit = y[X[f] <= t]
            rsplit = y[X[f] > t]

            lp = len(lsplit[lsplit == True]) / len(lsplit)
            lent = _compute_binary_entropy(lp)

            rp = len(rsplit[rsplit == True]) / len(rsplit)
            rent = _compute_binary_entropy(rp)

            ent = (len(lsplit) * lent + len(rsplit) * rent) / len(y)
            return ent
      
        best_ent = 10.0
        best_feature = None
        best_split = None
      
        for f in X.columns:
            values = X[f].sort_values().unique()
            for i, v in enumerate(values[:-1]):
                t = (v + values[i + 1]) / 2
                entropy = _compute_split_entropy(f, t)
                if entropy < best_ent:
                    best_ent = entropy
                    best_feature = f
                    best_split = t

        self.split_feature = best_feature
        self.split_value = best_split
        
        return [self.split_feature, self.split_value]
    
    def predict(self,X):
        return X[self.split_feature] <= self.split_value
      
ds = decision_stump()
ds.fit(train.loc[:,features], train.loc[:,target])
```

```python
'''
Compare your implementation to one from sklearn; in particular, fit a sklearn's
decision stump on the same training data, visualize the resulting tree and compare
to your model in terms of the splitting feature and the splitting value. Comment
on your findings.
'''

sk_ds = DTC(criterion='entropy', max_depth=1, random_state=43)
sk_ds.fit(train.loc[:,features], train.loc[:,target])

visualize(sk_ds)
```

As we see in the visualization above, and the feature split found before, both of them split on the feature "trump", at value 74.5.

```python
'''
Create another table like 'results_df' for the test set with the following rows:
(a) truth, (b) decision tree correct?, (c) decision stump correct? Does your decision
stump perform worse than the tree you fitted before? Are mistakes the same or not?
Comment on advantages and disadvatages of a decision stump compared to a decision tree.
'''

results_df = pd.DataFrame({
  'truth': test.loc[:,target],
  'model': model.predict(test.loc[:,features]),
  'ds': ds.predict(test.loc[:,features])
})
results_df['decision tree correct?'] = results_df.apply(lambda row: not (row['model'] ^ row['truth']), axis=1)
results_df['decision stump correct?'] = results_df.apply(lambda row: not (row['ds'] ^ row['truth']), axis=1)

results_df = results_df.drop(columns=['model', 'ds']).T

results_df
```

We note in the table above that the decision stump makes more mistakes, for CA and DE.

Decision stump is a decision tree with depth 1. As a result, it provides a very coarse split of the input space. Decision trees improve this by allowing more granular splits of the input space for decision making. An advantage of decision stumps is that they are very cheap to train, and at also at test time.

```python
# Part (c):
from sklearn.ensemble import RandomForestClassifier as RFC
'''
Fit a random forest of 200 sklearn's decision stumps. Use entropy as splitting
criterion, bootstrap sampling of the training data and consider a random subset
of only 5 features for each split (effectively, for each tree). 
'''

model = RFC(n_estimators=200, criterion='entropy', bootstrap=True, max_features=5, random_state=43)
model.fit(train.loc[:,features], train.loc[:,target])
```

```python
'''
Add the 'random forest correct?' row to the table we created in the previous part
How does it perform compared to the decision stump? Compared to the decision tree?
'''

results_df = results_df.T
results_df['rf'] = model.predict(test.loc[:,features])
results_df['random forest correct?'] = results_df.apply(lambda row: not (row['rf'] ^ row['truth']), axis=1)
results_df = results_df.drop(columns=['rf']).T

results_df
```

We see that the random forest classifier gives the correct results for CA and DE. It does, however, fail on NV.

```python
'''
Create a Pandas Dataframe showing one sample standard deviation of predictions across
all learners in the forest for each test state (hint: we did it in lab 2). Add the 
'random forest correct?' row to the table. Comment on your findings.
'''

predictions=[]
for tree in model.estimators_:
    predictions.append(tree.predict(test.loc[:,features]).tolist())
predictions = np.array(predictions).T
sigma_hat = np.sqrt(np.sum((predictions - np.expand_dims(test.loc[:,target], axis=-1))**2, axis=-1) / (predictions.shape[-1] - 1))

results_df = results_df.T
results_df['standard deviation'] = sigma_hat
results_df = results_df.T

results_df
```

We note that some of the highest standard deviations happen for the erroneous predictions like MI, NV and PA. This is intuitively what we would expect.
