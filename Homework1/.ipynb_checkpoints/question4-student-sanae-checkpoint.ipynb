```python
import numpy as np
import pandas as pd
train=pd.read_csv('train_elections.csv',index_col=0)
test=pd.read_csv('test_elections.csv',index_col=0)
features=train.columns[:-2]
target="trumpWinner"
```

```python
# Part (a):

import graphviz
import pydotplus
from sklearn.tree import export_graphviz as export
from sklearn.tree import DecisionTreeClassifier as DTC

def visualize(skl_tree):
    dot_format=export(skl_tree,class_names=["Clinton","Trump"],filled=True,feature_names=features)
    pydot_graph=pydotplus.graph_from_dot_data(dot_format)
    pydot_graph.set_size('"4,4!"')
    display(graphviz.Source(pydot_graph.to_string()))
    
'''Fit a decision tree on your train data and visualize it'''

# definiting the tree model
model=DTC(criterion='entropy', min_samples_split=5)
# fitting the model on the training data
model.fit(train.loc[:,features],train.loc[:,target])

visualize(model)
```

```python
'''
Recreate this tree using if-else statements by completing the below function.
 - argument: a Pandas DataFrame corresponding to a single test state;
 - return: a boolean prediction for the target variable;
'''

def tree_alias(inp):
    '''Function that recreate the decision tree, as plotted above
    and shown in the pdf version of answers (please refer to the latter
    as the decision tree changes one its split features randomly)
    '''
    if inp["trump"] <= 74.5:
        if inp["gun"] <= 63.0:
            return False
        else:
            return True
    elif inp["gun"] <= 80.5:
        if inp["obamacare"] <= 91.0:
            return False
        else:
            return True
    else:
        return True
```

```python
ifelse_preds=[tree_alias(test.loc[state]) for state in test.index]
sklearn_preds=model.predict(test.loc[:,features])
results=[[int(ie),int(sk),int(not ie^sk)] for ie,sk in zip(ifelse_preds,sklearn_preds)]
results_df=pd.DataFrame(results,columns=['tree alias','tree sklearn','match?'],index=test.index)
display(results_df.T)
```

```python
'''
Fetch predictions from the fitted model (or use your 'tree_alias function')
and compare them to ground truth on the test set. In particular, create a Pandas
DataFrame named 'results_df' as above but with 'predicted', 'truth', 'correct?'
as rows and test states as columns. How many are classified correctly?
'''

preds=model.predict(test.loc[:,features])
results=[[int(p),int(t),int(not p^t)] for p,t in zip(preds,test.loc[:,target])]
results_df=pd.DataFrame(results,columns=['predicted','truth','correct?'],index=test.index)
display(results_df.T)
print("Number of correct classifications: {}".format(sum(results_df["correct?"])))
```

```python
# Part (b):

import math
def extended_log(arg):
    return 0 if not arg else math.log(arg)
def entropy(p):
    return -sum([p_i*extended_log(p_i) for p_i in p])

'''
Complete the class 'decision stump' that has attributes/methods:
- attribute 'self.split_feature': the name of the optimal splitting feature;
- attribute 'self.split_value': value at which self.split_feature should be split on;
- method 'fit(self,X,y)': decides on the above two attributes based on given Pandas
DataFrames X and y holding features and targets for all training examples, respectively.
Returns the list [self.split_feature,self.split_value] after fitting.
- method 'predict(self,X)': returns a list of predictions given a Pandas DataFrame X
holding features of some number of test examples. 
'''

class decision_stump():
    ''' Implementation of a decision stump
    '''
    def __init__(self):
        self.split_feature=None
        self.split_value=None
        self.dec_lnode=None
        self.dec_rnode=None
        
    def fit(self,X,y):
        min_w_entropy = float("inf")
        split_feature = None
        l = len(X)
        features_names = X.columns
        t_name = y.columns[0]
        X = pd.concat([X, y], axis=1, sort=False)
        for f_name in features_names:
            mid_values = (np.sort(np.unique(X[f_name]))[:-1] + np.sort(np.unique(X[f_name]))[1:])/2
            for value in mid_values:
                l_node = X[X[f_name] <= value]
                r_node = X[X[f_name] > value]
                p_lnode_t = len(l_node[l_node[t_name] == True]) / len(l_node)
                p_rnode_t = len(r_node[r_node[t_name] == True]) / len(r_node)
                l_node_ent = entropy([p_lnode_t, 1-p_lnode_t])
                r_node_ent = entropy([p_rnode_t, 1-p_rnode_t])
                weighted_entropy = len(l_node)/l * l_node_ent + len(r_node)/l * r_node_ent
                if weighted_entropy < min_w_entropy:
                    min_w_entropy = weighted_entropy
                    self.split_feature = f_name
                    self.split_value = value
        l_node = X[X[self.split_feature] <= self.split_value]
        r_node = X[X[self.split_feature] > self.split_value]
        self.dec_lnode = max(set(list(l_node[t_name])), key = list(l_node[t_name]).count) 
        self.dec_rnode = max(set(list(r_node[t_name])), key = list(r_node[t_name]).count) 
        
        return [self.split_feature, self.split_value]
    
    def predict(self,X):
        
        return [self.dec_lnode if i <= self.split_value else self.dec_rnode for i in X[self.split_feature]]
```

```python
# Part (c):
from sklearn.ensemble import RandomForestClassifier as RFC
'''
Fit a random forest of 200 sklearn's decision stumps. Use entropy as splitting
criterion, bootstrap sampling of the training data and consider a random subset
of only 5 features for each split (effectively, for each tree). 
'''
model_rf=RFC(n_estimators=200,criterion='entropy',max_depth=1,max_features=5,bootstrap=True, random_state=43)
model_rf.fit(train.loc[:,features],train.loc[:,target])
```

```python
'''
Add the 'random forest correct?' row to the table we created in the previous part
How does it perform compared to the decision stump? Compared to the decision tree?
'''
preds_rf=model_rf.predict(test.loc[:,features])
results_df["random forest correct?"]=[int(not p^t) for p,t in zip(preds_rf,test.loc[:,target])]
display(results_df.T)
```

```python
'''
Create a Pandas Dataframe showing one sample standard deviation of predictions across
all learners in the forest for each test state (hint: we did it in lab 2). Add the 
'random forest correct?' row to the table. Comment on your findings.
'''

# retrieve individual predictions of the trees
predictions=[]
for tree in model_rf.estimators_:
    predictions.append(tree.predict(test.loc[:,features]))
    
# define a function that returns the uncertainty
def uncertainty(x,y,y_hat):
    '''Returns the incertainty as defined in the notebook from lab 2'''
    y_hat_x=[prediction[x] for prediction in y_hat]
    sigma_hat=np.sqrt(np.divide(sum([(p-y[x])**2 for p in y_hat_x]),199))
    return sigma_hat

results_df["uncertainty for random forest"] = [round(uncertainty(obs,test.loc[:,target],predictions),3) for obs in range(len(test))]
display(results_df.T)
```
