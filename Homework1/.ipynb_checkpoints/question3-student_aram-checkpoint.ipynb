```python
import numpy as np
import pandas as pd
train=pd.read_csv('train_elections.csv',index_col=0)
test=pd.read_csv('test_elections.csv',index_col=0)
features=train.columns[:-2]
target="trumpWinner"
```

```python
# Part (a):

import graphviz
import pydotplus
from sklearn.tree import export_graphviz as export
from sklearn.tree import DecisionTreeClassifier as DTC

def visualize(skl_tree):
    dot_format=export(skl_tree,class_names=["Clinton","Trump"],filled=True,feature_names=features)
    pydot_graph=pydotplus.graph_from_dot_data(dot_format)
    pydot_graph.set_size('"4,4!"')
    display(graphviz.Source(pydot_graph.to_string()))

'''Fit a decision tree on your train data and visualize it'''

tree_a = DTC(criterion='entropy',min_samples_split=5)
tree_a.fit(train.loc[:,features],train.loc[:,target])

visualize(tree_a)
```

```python
'''
Recreate this tree using if-else statements by completing the below function.
 - argument: a Pandas DataFrame corresponding to a single test state;
 - return: a boolean prediction for the target variable;
'''

def tree_alias(inp):
    if inp["trump"] <= 74.5:
        if inp["gun"] <= 63.0:
            return 0
        else:
            return 1
    else:
        if inp["immigration"] <= 43.0:
            return 1
        else:
            if inp["obamacare"] <= 91.0:
                return 0
            else:
                return 1
```

```python
# Check your results (run this cell):
###Comparing Tree Alias versus SkLearn tree

#generate predictions from tree_alias
ifelse_preds=[tree_alias(test.loc[state]) for state in test.index]
#generate predictions from sklearn tree
sklearn_preds=tree_a.predict(test.loc[:,features])
#compare
results=[[int(a),int(b),int(not a^b)] for a,b in zip(ifelse_preds,sklearn_preds)]
#make dataframe and display
results_df_a1=pd.DataFrame(results,columns=['Tree alias preciction','Sklearn Tree prediction','match?']
                           ,index=test.index)
display(results_df_a1.T)
```

```python
# Part (b):

import math
def extended_log(arg):
    return 0 if not arg else math.log(arg)
def entropy(p):
    return -sum([p_i*extended_log(p_i) for p_i in p])

'''
Complete the class 'decision stump' that has attributes/methods:
- attribute 'self.split_feature': the name of the optimal splitting feature;
- attribute 'self.split_value': value at which self.split_feature should be split on;
- method 'fit(self,X,y)': decides on the above two attributes based on given Pandas
DataFrames X and y holding features and targets for all training examples, respectively.
Returns the list [self.split_feature,self.split_value] after fitting.
- method 'predict(self,X)': returns a list of predictions given a Pandas DataFrame X
holding features of some number of test examples. 
'''

class decision_stump():
    def __init__(self):
        self.split_feature=None
        self.split_value=None
    def fit(self,X,y):
        TotalLen = len(X)
        feats = X.columns
        targs = y.columns[0]
        Xcat = pd.concat([X,y],axis=1,sort=False)
        
        minEnt = np.inf #set entropy to maximum value
        
        for f in feats:
            mid = ( np.sort(np.unique(Xcat[f]))[:-1] + np.sort(np.unique(Xcat[f]))[1:] )/2
        
            for v in mid:
                
                #compute left and right side of node
                Lbool = Xcat[f] <= v
                Rbool = Xcat[f] > v 
                Left = Xcat[Lbool]
                Right = Xcat[Rbool]
                
                #compute probabilities for left and right side
                Ltarg = Left[targs] == True
                Rtarg = Right[targs] == True
        
                ProbLeft = len(Left[Ltarg])/len(Left)
                ProbRight = len(Right[Rtarg])/len(Right)
        
                #compute entropies using the above probabilities
                EntLeft = entropy([ProbLeft,1-ProbLeft])
                EntRight = entropy([ProbRight,1-ProbRight])
        
                #compute expected/weighted entropy
                ExpectedEnt = (len(Left)/TotalLen)*EntLeft + (len(Right)/TotalLen)*EntRight
            
                #compare to running min
                if ExpectedEnt < minEnt:
                    #if better, then use this feature/value
                    minEnt = ExpectedEnt
                    self.split_feature=f
                    self.split_value=v
                    
        #final decision output
        LBoolEnd = Xcat[self.split_feature]<=self.split_value
        RBoolEnd = Xcat[self.split_feature]>self.split_value

        LeftEnd = Xcat[LBoolEnd]
        RightEnd = Xcat[RBoolEnd]
        
        self.LeftDecision = max(set(list(LeftEnd[targs])),  key=list(LeftEnd[targs]).count )
        self.RightDecision = max(set(list(RightEnd[targs])),  key=list(RightEnd[targs]).count )
        
        return [self.split_feature,self.split_value]
    def predict(self,X):
        
        return [self.LeftDecision if i<= self.split_value else self.RightDecision for i in X[self.split_feature]]
```

```python
dec_stump = decision_stump()
split_feat, split_val = dec_stump.fit(train[features],train[[target]])
print("Optimal feature split: '{}'".format(split_feat))
print("Optimal feature split value: '{}'".format(split_val))

```

```python
'''
Compare your implementation to one from sklearn; in particular, fit a sklearn's
decision stump on the same training data, visualize the resulting tree and compare
to your model in terms of the splitting feature and the splitting value. Comment
on your findings.
'''

tree_stump = DTC(criterion='entropy',max_depth=1)
tree_stump.fit(train.loc[:,features],train.loc[:,target])

visualize(tree_stump)
```

```python
'''
Create another table like 'results_df' for the test set with the following rows:
(a) truth, (b) decision tree correct?, (c) decision stump correct? Does your decision
stump perform worse than the tree you fitted before? Are mistakes the same or not?
Comment on advantages and disadvatages of a decision stump compared to a decision tree.
'''
#get ground truth
truth=[(test.loc[state])['trumpWinner'] for state in test.index]

#prediction output from tree_stump I wrote
dec_stump_preds = dec_stump.predict(test)

#prediction output from sklearn stump
sklearn_preds_stump=tree_stump.predict(test.loc[:,features])

#prediction output from sklearn tree (from part (A))
sklearn_preds_tree=tree_a.predict(test.loc[:,features])


results=[[int(a),int(not a^b), int(not a^c)] 
         for a,b,c in zip(truth,sklearn_preds_tree,sklearn_preds_stump)]
results_df_b1=pd.DataFrame(results,columns=['truth','Sklearn Tree correct?','Sklearn Stump correct?']
                           ,index=test.index)
display(results_df_b1.T)
```

Sklearn Stump had 4 incorrect whereas Sklearn Tree had 3 wrong; 2 of the 4 match those of the tree.  

Decision stumps are faster and cheaper to evaluate, and less prone to overfitting than standard decision trees. The downside is they can also vary in their prediction (i.e. classic bias-variance trade-off). Stumps are often referred to as "weak" learners, but when aggregated via boosting (i.e. AdaBoost), they can lead to strong predictions.

```python
# Part (c):
from sklearn.ensemble import RandomForestClassifier as RFC
'''
Fit a random forest of 200 sklearn's decision stumps. Use entropy as splitting
criterion, bootstrap sampling of the training data and consider a random subset
of only 5 features for each split (effectively, for each tree). 
'''

model_rando = RFC(n_estimators=200,criterion='entropy',max_depth=1,max_features=5,bootstrap=True)
model_rando.fit(train.loc[:,features],train.loc[:,target])
```

```python
'''
Add the 'random forest correct?' row to the table we created in the previous part
How does it perform compared to the decision stump? Compared to the decision tree?
'''

#get ground truth
truth=[(test.loc[state])['trumpWinner'] for state in test.index]

#prediction output from sklearn stump (from part (B))
sklearn_preds_stump=tree_stump.predict(test.loc[:,features])

#prediction output from sklearn tree (from part (A))
sklearn_preds_tree=tree_a.predict(test.loc[:,features])

#prediction output from sklearn random forest
rando_predictions = model_rando.predict(test.loc[:,features])

results=[[int(a),int(not a^b), int(not a^c), int(not a^d)] 
         for a,b,c,d in zip(truth,sklearn_preds_tree,sklearn_preds_stump, rando_predictions)]
results_df_c1=pd.DataFrame(results,columns=['truth','Sklearn Tree correct?','Sklearn Stump correct?','Sklearn RF correct?'],
                           index=test.index)
display(results_df_c1.T)
```

```python
'''
Create a Pandas Dataframe showing one sample standard deviation of predictions across
all learners in the forest for each test state (hint: we did it in lab 2). Add the 
'random forest correct?' row to the table. Comment on your findings.
'''

predictions=[]
for tree in range(200):
    predictions.append(model_rando.estimators_[tree].predict(test.loc[:,features]))
    
def uncertainty(x,y,yhat):
    yhat_x = [prediction[x] for prediction in yhat]
    sigmahat = np.sqrt(np.divide(sum([(p-y[x])**2  for p in yhat_x]),200))
    return sigmahat

#get ground truth
truth=[(test.loc[state])['trumpWinner'] for state in test.index]

#prediction output from sklearn stump (from part (B))
sklearn_preds_stump=tree_stump.predict(test.loc[:,features])

#prediction output from sklearn tree (from part (A))
sklearn_preds_tree=tree_a.predict(test.loc[:,features])

#prediction output from sklearn random forest
rando_predictions = model_rando.predict(test.loc[:,features])

results=[[int(a),int(not a^b), int(not a^c), int(not a^d)] 
         for a,b,c,d in zip(truth,sklearn_preds_tree,sklearn_preds_stump, rando_predictions)]
results_df_c2=pd.DataFrame(results,columns=['truth','Sklearn Tree correct?','Sklearn Stump correct?','Sklearn RF correct?'],
                           index=test.index)

results_df_c2['uncertainty'] = [round(uncertainty(obs,test.loc[:,target],predictions),3) 
                                for obs in range(len(test))]

display(results_df_c2.T)
```

We note that the RandomForest has also four mistakes, corresponding to an uncertainty level above roughly 0.690 or so. Of these 4 errors, 3 are the ones that occurred from the naive 1-stump approach; this difficulty must have carrier over at a large enough scale to affect the random forest with 200 learners.
