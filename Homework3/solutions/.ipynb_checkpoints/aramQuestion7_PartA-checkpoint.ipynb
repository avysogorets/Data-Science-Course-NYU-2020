```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans as km
from sklearn.cluster import AgglomerativeClustering as ag
from auxiliary import *
```

```python
# The linearly non-separable example:
A = [0.,0.]
B = [0.,1.]
C = [0.,3.]
D = [2.5,1.]
E = [3.,1.]
F = [-1., 2.]
G = [4.,2.]

data1= np.array([A,B,C,D,E,F,G])
plt.scatter(data1[:,0],data1[:,1],color='lightskyblue',alpha=0.8)
#plt.show()

#model_max=ag(n_clusters=2,affinity='manhattan',linkage='complete')
#preds_max=model_max.fit_predict(data1)
#plt.scatter(data1[:,0],data1[:,1],color=['orange' if p>0 else 'purple' for p in preds_max],s=50)
names = [' A', ' B', ' C', ' D', ' E',' F',' G']
for i, txt in enumerate(names):
    plt.annotate(txt, (data1[i,0], data1[i,1]))
plt.show()
```

```python
from scipy.cluster import hierarchy
Z = hierarchy.linkage(data1, method='complete',metric='cityblock')

fig = plt.figure(figsize=(25, 10))
dn = hierarchy.dendrogram(Z)
plt.show()
```
