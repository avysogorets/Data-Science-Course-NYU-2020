### *K-Means Clustering*
Given a set of points $\mathcal{D}=\{x_1,x_2,\ldots,x_n\}\subset\mathbb{R}^d$, we seek to obtain $k$ centroids $m_1,m_2,\ldots,m_k\in\mathbb{R}^d$ as to minimize the sum of squared distances from each point to the nearest centroid: $\phi(\bar{m})=\sum_{i=1}^{n}\min_{j\in[k]}\lVert m_j-x_i\rVert^2$ (more generally, squared Euclidean distance can be replaced by any other squared metric on $\mathbb{R}^n$ such as cosine similarity). This problem is NP-hard, but a well-known iterative algorithm applies to yield locally optimal solutions:

 0. select initial centroids $m_1^{(0)},m_2^{(0)},\ldots,m_k^{(0)}$ uniformly from $\mathcal{D}$;
 1. assign points in $\mathcal{D}$ to the nearest centroid, i.e. let $C_j^{(t)}=\{x_i\in\mathcal{D}\colon j=\text{argmin}_{l\in[k]}\lVert x_i-m_l\rVert\}$ be the set of points having $m_j^{(t)}$ as the closest centroid;
 2. update centroids by averaging points assigned to them: $m_j^{(t+1)}=\left|C_j^{(t)}\right|^{-1}\sum_{x_i\in C_j^{(t)}}x_i$;
 3. iterate steps 1 and 2 until convergence (until $C_j^{(t)}\ne C_j^{(t+1)}$ for some $j\in[k]$).

Why does this algorithm converge? To see this, note that the cost function $\phi(\bar{m})$ does not increase after each iteration. Indeed, let $\bar{m}=(m_1,m_2,\ldots,m_k)$ for brevity and $s_i(\bar{m})=\text{argmin}_{j\in[k]}\lVert x_i-m_j\rVert$ be the index of the closest centroid to observation $x_i$. Finally, define $Q(\bar{m}',\bar{m})=\sum_{i=1}^{n}\lVert x_i-m'_{s_i(\bar{m})}\rVert$ to be the sum of distances from each point to its assigned centroid $m'_{s_i(\bar{m})}$ under the partition induced by $\bar{m}$. Straight from definitions, we clearly see that $Q$ *majorizes* $\phi$, i.e. $Q(\bar{m},\bar{m})=\phi(\bar{m})$ and $Q(\bar{m}',\bar{m})\geq\phi(\bar{m}')$ for any $\bar{m}',\bar{m}\in\mathbb{R}^d$. Letting $\bar{m}=\bar{m}^{(t)}$ and $\bar{m}'=\bar{m}^{(t+1)}$, we also have $Q(\bar{m}^{(t+1)},\bar{m}^{(t)})\leq Q(\bar{m}^{(t)},\bar{m}^{(t)})$ because given a fixed set of points, their average minimizes the sum of squared distances to these points among all other points in $\mathbb{R}^d$ (we will prove it shortly). In this case, combining all identities and inequalities, we have

\begin{align}
\phi\left(\bar{m}^{(t+1)}\right)-\phi\left(\bar{m}^{(t)}\right)=\phi\left(\bar{m}^{(t+1)}\right)-Q\left(\bar{m}^{(t+1)},\bar{m}^{(t)}\right)+Q\left(\bar{m}^{(t+1)},\bar{m}^{(t)}\right)-Q\left(\bar{m}^{(t)},\bar{m}^{(t)}\right)\leq 0,
\end{align}

which means that $\phi(\bar{m}^{(t+1)})\leq\phi(\bar{m}^{(t)})$, so that the error is indeed non-increasing over iterations. Finally, to complete the proof, we need to show that given some points $z_1,z_2,\ldots,z_n$, we have that $c(y)=\sum_{i=1}^{n}\lVert z_i-y\rVert^2$ attains minimum for $y=n^{-1}\sum_{i=1}^{n}z_i$. Note that $c(y)$ is convex in all coordinates of $y$, so the $j$-th component of the optimal $y$ is the root of $\partial c(y)/\partial y_j=0$:

\begin{align}
\frac{\partial c(y)}{\partial y_j}=\sum_{i=1}^{n}2(z_{ij}-y_j)=0\:\:\text{ is equivalent to}\:\:\:y_j=\frac{1}{n}\sum_{i=1}^{n}z_{ij},
\end{align}
completing the proof.

 4. K-Means does not have any performance guarantees; its solution can be arbitrarily worse than the optimum. 

Let $\phi_{OPT}$ denote the minimum possible cost for given data points $\mathcal{D}$ and the number of centroids $k$. Then, it is argued that for any $\epsilon>0$, there exist such $\mathcal{D}$ and $k$ that the expected final cost $\mathbb{E}\phi$ will satisfy $\phi_{OPT}^{-1}\mathbb{E}\phi>\epsilon$ (the expectation is taken over uniform initializations of centroids). To see this, consider a dataset $\mathcal{D}\subset\mathbb{R}^2$ of $3n$ points that are equally distributed among 3 different clusters or radii $\delta$ whose centers are separated by $B+2\delta>9\delta$, as shown below:

Then, in particular, assuming $n$ fixed, there is some positive probability $p$ that initial centroids will be distributed as follows:

Consider the progression of 3-means algorithm for the above situation. All points from the leftmost cluster will clearly be assigned to the leftmost centroid; the same is true for points in the middle cluster. Points from the rightmost cluster will be somehow split between the two centroids in their neighborhood. Thus, after the first update, we expect

where the leftmost centroid, assigned to the mean of all $2n$ points in the leftmost and middle clusters, can be anywhere inside the specified circle (depending on how points are distributed inside both clusters). Note that we need $B>7\delta$ to ensure that neither of the two centroids in the rightmost cluster can ever become closer to any point in the middle cluster than the leftmost centroid. Now, the leftmost cluster will hold in its exact position inside the circle as no futher reassignment of points from the first two clusters will occur. Some perturbation is possible in the rightmost cluster, however, neither of the two centroids there will ever escape the $\delta$-neighborhood and thus will not affect anything on the outside.

Now, focusing on the leftmost centroid alone, we see that the optimal error for this initial configuration cannot be smaller than $0.25nB^2$, which is a lower bound on the sum of squared distances from this centroid to points in either left or middle cluster (whichever is futher away from it). Thus, for this $\mathcal{D}$ and $k=3$, $\mathbb{E}\phi\geq 0.25npB^2$, whereas $\phi_{OPT}\leq 3n\delta^2$ (when each centroid is placed in the center of a corresponding cluster). Hence, $\phi_{OPT}^{-1}\mathbb{E}\phi\geq 12^{-1}p(B/\delta)^{2}$. Now, given any $\epsilon$, it is possible to find $B,\delta$ such that this ratio is no less than $\epsilon$, proving the claim.

### *K-Means++*
This extension to the above algorithm claims to find competitive initial centroid seeds, which happen to provide performace guarantees that the standard K-means lacks. The seeding procedure is as follows:

 0. select a point uniformly from $\mathcal{D}$ and let it be the position of the first centroid $c_1$;
 1. assume centroids $c_1,c_2,\ldots, c_m$ are chosen; the next centroid $c_{m+1}$ is drawn from $\mathcal{D}$ with $\mathbb{P}(c_{m+1}=x_i)\sim\min_{j\in[m]}\lVert x_i-c_m\rVert^2$, i.e., the probability of having $c_{m+1}=x_i$ is the normalized squared distance from $x_i$ to the closest centroid among $c_1,c_2,\ldots,c_m$.
 2. Repeat the above step until $m=k$;
 3. run the K-means algorithm with the selected initial centroids.
 
Scikit-learn implements K-means++; one simply needs to set the ```init``` variable to ```k-means++``` when instantiating the K-means class (in fact,  ```init``` defaults to ```k-means++```, so the vanilla K-means algorithm is actually K-means++).

[The authors of K-means++ showed](http://ilpubs.stanford.edu:8090/778/1/2006-13.pdf) that $\mathbb{E}\phi\leq(8\log k +2)\phi_{OPT}$ already after initialization! The proof is rather lengthly; you can visit their original paper for the full derivation. 

````python
# The "sensitivity" example revisited:
fig,axes=plt.subplots(nrows=4,ncols=3,figsize=(18,13))
costs=[km(n_clusters=4,init='random',algorithm='full',n_init=1).fit(data).inertia_ for _ in range(20)]
costspp=[km(n_clusters=4,init='k-means++',algorithm='full',n_init=1).fit(data).inertia_ for _ in range(20)]
data=np.array([[0,0],[0.2,-0.1],[0.7,0.85],[0.3,0.6],[0.65,0.8],[0.7,0],[0.3,-0.1],[-0.1,0.7],[0.3,0.05],[0.24,0.28],[0.1,0.37],[0.24,0.32],[0.2,0.57],[0.65,-0.12],[0.55,0],[0.74,-0.1],[0.8,0.05],[0.6,0.37],[0.7,0.43],[0.5,0.55]])
initials=np.array([data[np.random.choice(range(len(data)),size=4,replace=False)] for _ in range(4)])
for i,initial in enumerate(initials):
    model=km(n_clusters=4,init=initial,algorithm='full',n_init=1)
    model.fit(data)
    visualize(model,axes[i,1],color=False)
    modelpp=km(n_clusters=4,init='k-means++',algorithm='full',n_init=1)
    modelpp.fit(data)
    visualize(modelpp,axes[i,2],color=False)
    axes[i,0].scatter(initial[:,0],initial[:,1],facecolors='none',edgecolors='maroon',s=120)
    axes[i,0].scatter(initial[:,0],initial[:,1],facecolors='none',edgecolors='maroon',s=120)
    axes[i,0].scatter(data[:,0],data[:,1],color='lightskyblue',alpha=0.8)
    axes[i,0].set_xlim((-0.75,1.25))
    axes[i,0].set_ylim((-0.5,1.5))
    axes[i,1].scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='k',s=80,marker="$X$",label="centroids")
    axes[i,1].scatter(data[:,0],data[:,1],color='grey',alpha=0.3)
    axes[i,1].legend(loc='lower left')
    axes[i,1].text(x=-0.7,y=1.3,s=f"cost: {model.inertia_:.4f}")
    axes[i,2].scatter(modelpp.cluster_centers_[:,0],modelpp.cluster_centers_[:,1],color='k',s=80,marker="$X$",label="centroids")
    axes[i,2].scatter(data[:,0],data[:,1],color='grey',alpha=0.3)
    axes[i,2].legend(loc='lower left')
    axes[i,2].text(x=-0.7,y=1.3,s=f"cost: {modelpp.inertia_:.4f}")
print(f"4-Means average cost: {np.mean(costs):.4f}")
print(f"4-Means++ average cost: {np.mean(costspp):.4f}")
plt.show()
````

### *Global K-Means (GKM)*
This is yet another extension that aims to improve optimality of the original K-means algorithm. The procedure is as follows:

 0. initialize the first centroid $m_1^{*}=\frac{1}{n}\sum_{i=1}^{n}x_i$ to be the mean of all points in $\mathcal{D}$; note that $m_1^{*}$ is the optimal solution of $1$-Means problem;
 1. let $m^{*}_1,m^{*}_2,\ldots,m^{*}_{j-1}$ be the solution obtained on the previous step ($2\leq j\leq k$). For each point $x\in\mathcal{D}$, fit a $j$-Means algorithm with the following initial centroids: $m_i^{(0)}=m_i^{*}$
if $i<j$ and $m_j^{(0)}=x$. Among $n$ resulting fits, select one with the lowest error $\phi$.
 2. repeat step 2 until $j=k$.

As you can tell, GKM is much more computationally expensive: it requires $N(k-1)$ repetitions of the original K-means algorithm. Can you identify any source of stochasticity in this algorithm or is it deterministic?

````python
# Implementation of global K-Means using sklearn's regular K-Means:
class GlobalKMeans():
    def __init__(self,k):
        self.inertia_=None
        self.model=None
        self.centroids=None
        self.k=k
    def fit(self,data):
        if self.k==1:
            self.centroids=np.mean(data,axis=0)
        else:
            previous=GlobalKMeans(self.k-1)
            previous.fit(data)
            models={}
            centroids=previous.centroids
            candidates={}
            for point in data:
                model=km(n_clusters=self.k,init=np.vstack([centroids,point]),algorithm='full',n_init=1)
                model.fit(data)
                candidates[model.inertia_]=model.cluster_centers_
                models[model.inertia_]=model
            self.centroids=candidates[min(candidates)]
            self.model=models[min(models)]
            self.inertia_=self.model.inertia_
        return self.model
    def predict(self,data):
        return self.model.predict(data)
````

````python
# Compare all three K-Means algorithms:
costs=[km(n_clusters=4,init='random',algorithm='full',n_init=1).fit(data).inertia_ for _ in range(50)]
costspp=[km(n_clusters=4,init='k-means++',algorithm='full',n_init=1).fit(data).inertia_ for _ in range(50)]
costsg=[GlobalKMeans(k=4).fit(data).inertia_ for _ in range(5)]
print(f"global 4-means costs: {[np.round(cost,7) for cost in costsg]}")
print(f"\n4-means average cost: {np.round(np.mean(costs),4)}")
print(f"4-means++ average cost: {np.round(np.mean(costspp),4)}")
print(f"global 4-means cost: {np.round(np.mean(costsg),4)}")
````

### *Agglomerative Clustering*
Agglomerative clustering (also called *bottom-up*), together with Divisive clustering (also called *top-down*) constitute two main algorithms of Hierarchical clustering. In agglomerative clustering, one starts with all samples as individual clusters; then, until the desired number of clusters is reached (or unless some other stopping criteria is met), it fuses pairs of clusters based on some particular similarity measure. Note that agglomerative clustering is fundamentally different from K-means; here, the algorithm determines clusters of training data and not centroids of some hypothetical clusters. More so, while K-means can naturally attribute a new point to a particular cluster (using the index of the nearest centroid), there is no natural way to assign new observations to clusters in agglomerative clustering (```scikit-learn```'s implementation of agglomerative clustering does not have the ```predict``` function for this reason).

The key procedure in agglomerative clustering is the assessment of cluster similarity. How to determine which clusters to merge? There are a few common *linkage* functions. Let $C_1,C_2,\ldots,C_m$ be the current clusters. Then,

 1. ***Ward***: minimizes the sum of squared distances within clusters; i.e., a pair of clusters to merge will be those minimizing $\sum_{i=1}^{m}\sum_{x,y\in C_i}\lVert x-y\rVert^2$ the sum of all squared intra-cluster diferences. Note the similarity of this approach to the selection of optimal splits in decision trees;
 2. ***Maximum (complete) linkage***: this approach finds a pair of clusters that have the smallest maximum distance across all pairs of points from the two clusters; i.e. it merges clusters $C_i, C_j$ with the smallest value of $\max_{x\in C_i,y\in C_j}\lVert x-y\rVert$ across all $i\ne j$.
 3. ***Average linkage***: similarly to *maximum linkage*, finds a pair of clusters having minimal average distance between pairs of their points across all pairs of different clusters; i.e., it merges clusters $C_i,C_j$ with the smallest value of $\frac{1}{M_{ij}}\sum_{x\in C_i,y\in C_j}\lVert x-y\rVert$, where $M_{ij}=|C_i||C_j|$.
 4. ***Minimum (signle) linkage***: similarly to *maximum linkage*, this method merges two clusters having minimal minimum distance between pairs of their points across all pairs of different clusters; i.e., it merges clusters $C_i,C_j$ with the smallest value of $\min_{x\in C_i,y\in C_j}\lVert x-y\rVert$ across all $i\ne j$.
 
Going back to the linearly non-separable example we had before, which of the above linkage functions we need to use to separate the two obvious clusters?
